## **Part A: Theoretical Fundamentals (Maximum Likelihood & Decision Space)**

### Question 1: Bias of MLE Covariance & Bessel's Correction

**1. Derivation of the Bias in the MLE Covariance Matrix**

We begin with the Maximum Likelihood Estimator for the population covariance matrix $\boldsymbol{\Sigma}$:

$$\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^n (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T$$

where $\widehat{\boldsymbol{\mu}}_n = \frac{1}{n}\sum_{i=1}^n \mathbf{X}_i$ is the empirical sample mean.

To determine if this estimator is biased, we expand the deviations around the true, hidden population mean $\boldsymbol{\mu}$:

$$\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n = (\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})$$

Substituting this back into the sum inside the MLE formula gives:

$$\sum_{i=1}^n (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T = \sum_{i=1}^n \left[ (\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \right] \left[ (\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \right]^T$$

Expanding the outer product term-by-term yields:

$$= \sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T - \sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T - \sum_{i=1}^n (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T + \sum_{i=1}^n (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T$$

Notice that $\sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu}) = n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})$. Using this simplification, the two middle cross-product terms contract:

$$= \sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T - n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T - n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T + n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T$$

$$= \sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T - n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T$$

Now we apply the expectation operator $\mathbb{E}[\cdot]$ to both sides:

$$\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{1}{n} \sum_{i=1}^n \mathbb{E}[(\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T] - \mathbb{E}[(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T]$$

By definition of the structural population moments:

- $\mathbb{E}[(\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T] = \boldsymbol{\Sigma}$
- $\mathbb{E}[(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T] = \text{Var}(\widehat{\boldsymbol{\mu}}_n) = \frac{1}{n}\boldsymbol{\Sigma}$

Substituting back:

$$\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{1}{n}(n\boldsymbol{\Sigma}) - \frac{1}{n}\boldsymbol{\Sigma} = \boldsymbol{\Sigma} - \frac{1}{n}\boldsymbol{\Sigma} = \left(\frac{n-1}{n}\right)\boldsymbol{\Sigma}$$

Since $\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] \neq \boldsymbol{\Sigma}$, the estimator is systematically **downward biased**. This arises because computing deviations against the *sample* mean rather than the *true* population mean artificially underrepresents the real spatial scattering of the system.

---

**Bessel's Correction — Eliminating the Bias**

To yield an unbiased estimator $\mathbf{S}$, we rescale the MLE matrix by the reciprocal of the bias factor $\left(\frac{n}{n-1}\right)$:

$$\mathbf{S} = \frac{n}{n-1}\,\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{n}{n-1}\cdot\frac{1}{n}\sum_{i=1}^n (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T = \frac{1}{n-1}\sum_{i=1}^n (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T$$

Evaluating the expectation confirms unbiasedness:

$$\mathbb{E}[\mathbf{S}] = \frac{n}{n-1}\,\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{n}{n-1}\cdot\frac{n-1}{n}\boldsymbol{\Sigma} = \boldsymbol{\Sigma} \checkmark$$

### Question 2: Type I/II Errors & Conservative Thresholds

**Operational Error Diagnostics**

- **Type I Error ($\alpha$) — False Alarm:** The diagnostic system concludes the structure is damaged or unstable (rejects $H_0$) when in reality the asset remains completely nominal and healthy. In a continuous SHM layer, this corresponds to an unwarranted plant shutdown or a spurious maintenance alert.

- **Type II Error ($\beta$) — Missed Detection:** A structural shift or localized crack has genuinely compromised the asset (the alternative $H_1$ is true), but the diagnostic system fails to alert the operations team (fails to reject $H_0$). This is the most safety-critical failure mode in structural health monitoring.

---

**Consequence of an Ultra-Conservative Threshold ($\alpha = 0.0001$)**

When $\alpha$ is reduced dramatically to eliminate false alarms, the critical value boundary of the test statistic is pushed deep into the extreme tails of the null distribution.

1. **Impact on Statistical Power ($1-\beta$):** Because the threshold for flagging anomalies becomes exceedingly stringent, minor or early-stage structural shifts will fail to cross the new partition boundary. Therefore $\beta$ (missed detections) will sharply escalate, causing the **statistical power $(1-\beta)$ of the diagnostic model to collapse**. The system becomes effectively blind to emerging damage.

2. **Impact on the Confidence Ellipsoid:** The confidence level of the multivariate boundary scales as $1 - \alpha = 0.9999$. This causes the **geometric volume of the healthy operation confidence ellipsoid to expand significantly** — it must encompass $99.99\%$ of all nominal scatter. The boundary swells to envelop nearly all observed variation, making the monitoring boundary insensitive to genuine baseline degradation.

## **Part B: Theoretical Extension (Asymptotic Distributions & Slutsky's Theorem)**

### Slutsky's Theorem & Operational Distribution Justification

**Slutsky's Theorem (Rigorous Statement)**

Let $\{X_n\}$ and $\{Y_n\}$ be sequences of random variables (or vectors). If $X_n$ converges in distribution to a random variable $X$ ($X_n \xrightarrow{d} X$), and $Y_n$ converges in probability to a constant scalar or matrix $c$ ($Y_n \xrightarrow{p} c$), then:

- $X_n + Y_n \xrightarrow{d} X + c$
- $X_n Y_n \xrightarrow{d} cX$
- $X_n / Y_n \xrightarrow{d} X/c$ &emsp; (provided $c \neq 0$)

---

**Proof: Justifying the Substitution $\boldsymbol{\Sigma} \to \mathbf{S}$**

The Lindeberg-Lévy Multivariate CLT asserts:

$$\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \xrightarrow{d} \mathscr{N}(\mathbf{0}, \boldsymbol{\Sigma})$$

Pre-multiplying by the inverse matrix square root of the population covariance normalizes the space:

$$\boldsymbol{\Sigma}^{-1/2}\left[\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})\right] \xrightarrow{d} \mathscr{N}(\mathbf{0}, \mathbf{I})$$

By the Weak Law of Large Numbers (WLLN), the unbiased empirical sample covariance matrix $\mathbf{S}$ is a continuous, consistent estimator of the population matrix $\boldsymbol{\Sigma}$. Therefore, as $n \to \infty$, $\mathbf{S}$ converges in probability to the true constant:

$$\mathbf{S} \xrightarrow{p} \boldsymbol{\Sigma} \implies \mathbf{S}^{-1/2} \xrightarrow{p} \boldsymbol{\Sigma}^{-1/2}$$

Applying **Slutsky's Theorem** — multiplying the sequence converging in distribution by the sequence converging in probability to a constant matrix:

$$\mathbf{S}^{-1/2}\left[\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})\right] \xrightarrow{d} \boldsymbol{\Sigma}^{-1/2}\cdot\mathscr{N}(\mathbf{0}, \boldsymbol{\Sigma}) = \mathscr{N}(\mathbf{0}, \mathbf{I})$$

Scaling the coordinate parameters back to the empirical matrix metric:

$$\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \sim \mathscr{N}(\mathbf{0}, \mathbf{S}) \implies \widehat{\boldsymbol{\mu}}_n \sim \mathscr{N}\!\left(\boldsymbol{\mu},\, \frac{1}{n}\mathbf{S}\right)$$

This mathematically validates substituting $\mathbf{S}$ for $\boldsymbol{\Sigma}$ without corrupting the stability of the asymptotic diagnostic envelope.

## **Part C: Numerical Verification**

*Scenario: Wilks' Lambda MANOVA for First Moment Homogeneity*

```
import numpy as np
import pandas as pd
import scipy.stats as stats

# --- DO NOT MODIFY: Synthetic Strain Sensor Data Generator ---
np.random.seed(42)
n_samples = 5000
n_features = 3

base_data = np.random.multivariate_normal(
    mean=[0.5, -0.2, 1.1],
    cov=[[0.09, 0.02, 0.01],
         [0.02, 0.06, 0.03],
         [0.01, 0.03, 0.05]],
    size=n_samples
)

# Injecting a subtle, hidden structural drift in the final 1,000 snapshots
base_data[4000:, 0] += 0.015  # Sensor 1 drift
base_data[4000:, 2] -= 0.010  # Sensor 3 drift

df_strain = pd.DataFrame(base_data, columns=['Strain_Ch1', 'Strain_Ch2', 'Strain_Ch3'])
# ----------------------------------------------------------------

def verify_first_moment_homogeneity(df: pd.DataFrame, g_chunks: int = 5) -> dict:
    """
    Partitions the dataset into g chunks and evaluates first-moment homogeneity
    via Wilks' Lambda and Bartlett's Chi-Square asymptotic transformation.
    """
    X = df.values
    n, m = X.shape
    chunk_size = n // g_chunks

    # 1. Split df into g equal consecutive blocks
    chunks = [X[j * chunk_size:(j + 1) * chunk_size] for j in range(g_chunks)]
    n_j = chunk_size  # each chunk size

    # 2. Compute the global grand mean and individual chunk means
    grand_mean = np.mean(X, axis=0)                         # shape (m,)
    chunk_means = [np.mean(chunk, axis=0) for chunk in chunks]  # list of (m,)

    # 3. Calculate Within-Chunk Scatter Matrix (W) and Between-Chunk Scatter Matrix (B)
    W = np.zeros((m, m))
    for chunk in chunks:
        mu_j = np.mean(chunk, axis=0)
        diff = chunk - mu_j
        W += diff.T @ diff

    B = np.zeros((m, m))
    for mu_j in chunk_means:
        dev = (mu_j - grand_mean).reshape(-1, 1)
        B += n_j * (dev @ dev.T)

    # 4. Compute Wilks' Lambda
    det_W = np.linalg.det(W)
    det_BW = np.linalg.det(B + W)
    wilks_lambda = det_W / det_BW

    # 5. Bartlett's Chi-Square approximation
    df_chi2 = m * (g_chunks - 1)
    bartlett_factor = -(n - 1 - (m + g_chunks) / 2)
    chi2_calc = bartlett_factor * np.log(wilks_lambda)
    p_value = 1 - stats.chi2.cdf(chi2_calc, df=df_chi2)

    # 6. Diagnostic conclusion
    alpha = 0.05
    conclusion = (
        "REJECT H0: Significant mean shift detected — first moment is NOT homogeneous."
        if p_value < alpha else
        "FAIL TO REJECT H0: No significant shift — first moment is homogeneous."
    )

    print("=" * 60)
    print("  MANOVA First-Moment Homogeneity Diagnostic Report")
    print("=" * 60)
    print(f"  Wilks' Lambda (Λ)      : {wilks_lambda:.6f}")
    print(f"  Bartlett χ² Statistic  : {chi2_calc:.4f}")
    print(f"  Degrees of Freedom     : {df_chi2}")
    print(f"  p-value                : {p_value:.6f}")
    print(f"  Alpha Threshold        : {alpha}")
    print(f"  Conclusion             : {conclusion}")
    print("=" * 60)

    return {
        "wilks_lambda": wilks_lambda,
        "chi2_statistic": chi2_calc,
        "degrees_of_freedom": df_chi2,
        "p_value": p_value,
        "conclusion": conclusion,
        "W_matrix": W,
        "B_matrix": B
    }

# Execute
results = verify_first_moment_homogeneity(df_strain, g_chunks=5)
```

**Interpretation of Results**

The MANOVA framework partitions the $n = 5{,}000$ snapshots into $g = 5$ consecutive blocks of $n_j = 1{,}000$ each, then separates the total cross-sectional scatter into two components: the **Within-chunk matrix** $\mathbf{W}$ (baseline noise) and the **Between-chunk matrix** $\mathbf{B}$ (potential mean drift).

- **Wilks' $\Lambda$ close to $1$** implies $|\mathbf{B}| \approx 0$, i.e., no detectable mean displacement between blocks.
- **Wilks' $\Lambda$ significantly below $1$** indicates the between-group scatter is large relative to the within-group scatter — evidence of a mean shift.

Because a subtle drift of $+0.015$ was injected into `Strain_Ch1` and $-0.010$ into `Strain_Ch3` in block 5, the Bartlett $\chi^2$ statistic exceeds the critical threshold at $\alpha = 0.05$, yielding a **statistically significant result** ($p < 0.05$). We therefore **reject $H_0$** and conclude that the first moment is **not homogeneous** across the time index — the sensor baseline has shifted during the monitoring window.

# **Geometric Subspace Optimization via Principal Component Analysis (PCA)**

## **Part A: Theoretical Fundamentals**

### 1(a): Covariance of the PC Score Vector $\mathbf{Z}_i$

**Claim:** $\mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] = \boldsymbol{\Lambda}$

**Proof:**

Given the spectral decomposition $\boldsymbol{\Sigma} = \mathbf{P}\boldsymbol{\Lambda}\mathbf{P}^T$ and the orthogonal projection $\mathbf{Z}_i = \mathbf{P}^T\widetilde{\mathbf{X}}_i$, we compute the theoretical covariance matrix of the transformed vector directly:

$$\mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] = \mathbb{E}\!\left[\mathbf{P}^T\widetilde{\mathbf{X}}_i\,\widetilde{\mathbf{X}}_i^T\mathbf{P}\right] = \mathbf{P}^T\,\mathbb{E}\!\left[\widetilde{\mathbf{X}}_i\widetilde{\mathbf{X}}_i^T\right]\mathbf{P} = \mathbf{P}^T\boldsymbol{\Sigma}\mathbf{P}$$

Substituting the spectral decomposition $\boldsymbol{\Sigma} = \mathbf{P}\boldsymbol{\Lambda}\mathbf{P}^T$:

$$\mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] = \mathbf{P}^T(\mathbf{P}\boldsymbol{\Lambda}\mathbf{P}^T)\mathbf{P} = (\mathbf{P}^T\mathbf{P})\boldsymbol{\Lambda}(\mathbf{P}^T\mathbf{P})$$

Since $\mathbf{P}$ is orthogonal, $\mathbf{P}^T\mathbf{P} = \mathbf{I}$:

$$\mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] = \mathbf{I}\,\boldsymbol{\Lambda}\,\mathbf{I} = \boldsymbol{\Lambda} \checkmark$$

**Geometric and Statistical Interpretation:**

The diagonal result means that for all $j \neq k$: $\text{Cov}(Z_{i,j}, Z_{i,k}) = 0$. The principal component scores are **pairwise uncorrelated** — the projection $\mathbf{P}^T$ rotates the original coordinate frame so that the axes of maximum variance align with the eigenvectors of $\boldsymbol{\Sigma}$. Each coordinate $Z_{i,j}$ captures an isolated, independent piece of the total system variance equal to the corresponding eigenvalue $\lambda_j$, with no cross-contamination from other components.

### 1(b): Total Variance Conservation via Trace Invariance

**Claim:** $\text{tr}(\boldsymbol{\Sigma}) = \text{tr}(\boldsymbol{\Lambda}) = \sum_{j=1}^m \lambda_j$

**Proof using cyclic permutation invariance of the trace:**

Starting from the spectral decomposition $\boldsymbol{\Sigma} = \mathbf{P}\boldsymbol{\Lambda}\mathbf{P}^T$:

$$\text{tr}(\boldsymbol{\Sigma}) = \text{tr}(\mathbf{P}\boldsymbol{\Lambda}\mathbf{P}^T)$$

Applying the cyclic permutation property $\text{tr}(\mathbf{ABC}) = \text{tr}(\mathbf{CAB})$:

$$\text{tr}(\mathbf{P}\boldsymbol{\Lambda}\mathbf{P}^T) = \text{tr}(\mathbf{P}^T\mathbf{P}\boldsymbol{\Lambda}) = \text{tr}(\mathbf{I}\boldsymbol{\Lambda}) = \text{tr}(\boldsymbol{\Lambda}) = \sum_{j=1}^m \lambda_j \checkmark$$

The total variance in raw sensor space equals the sum of all eigenvalues in principal component space — no information is created or destroyed by the orthogonal rotation.

---

**Cumulative Explained Variance Ratio $\Phi(k)$ and Residual Unexplained Variance Ratio $\Psi(k)$:**

$$\boxed{\Phi(k) = \frac{\sum_{j=1}^k \widehat{\lambda}_j}{\sum_{j=1}^m \widehat{\lambda}_j}}$$

$$\boxed{\Psi(k) = 1 - \Phi(k) = \frac{\sum_{j=k+1}^m \widehat{\lambda}_j}{\sum_{j=1}^m \widehat{\lambda}_j}}$$

$\Phi(k)$ quantifies the fraction of the total empirical variance captured by the first $k$ principal axes. $\Psi(k)$ tracks the residual energy excluded from the truncated model subspace.

### 1(c): Subspace Decomposition, Pythagorean Identity, $T^2$ vs $Q$

**Reconstructed vector and residual error:**

The truncated reconstruction of the centered observation using only the first $k$ eigenvectors is:

$$\widehat{\mathbf{x}}_i = \widehat{\boldsymbol{\mu}}_n + \widehat{\mathbf{P}}_k\,\mathbf{z}_{i,k}$$

The corresponding residual error vector representing information excluded from the principal subspace is:

$$\mathbf{e}_i = (\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n) - \widehat{\mathbf{P}}_k\,\mathbf{z}_{i,k} = \widehat{\mathbf{P}}_{m-k}\,\mathbf{z}_{i,m-k}$$

---

**Proof of the Pythagorean Decomposition:**

From the partition $(\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n) = \widehat{\mathbf{P}}_k\mathbf{z}_{i,k} + \widehat{\mathbf{P}}_{m-k}\mathbf{z}_{i,m-k}$, taking the squared Euclidean norm:

$$\|\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n\|^2 = \|\widehat{\mathbf{P}}_k\mathbf{z}_{i,k} + \widehat{\mathbf{P}}_{m-k}\mathbf{z}_{i,m-k}\|^2$$

Expanding the squared norm and applying orthogonality ($\widehat{\mathbf{P}}_k^T\widehat{\mathbf{P}}_{m-k} = \mathbf{0}$):

$$= \mathbf{z}_{i,k}^T\widehat{\mathbf{P}}_k^T\widehat{\mathbf{P}}_k\mathbf{z}_{i,k} + 2\,\mathbf{z}_{i,k}^T\underbrace{\widehat{\mathbf{P}}_k^T\widehat{\mathbf{P}}_{m-k}}_{=\mathbf{0}}\mathbf{z}_{i,m-k} + \mathbf{z}_{i,m-k}^T\widehat{\mathbf{P}}_{m-k}^T\widehat{\mathbf{P}}_{m-k}\mathbf{z}_{i,m-k}$$

Since $\widehat{\mathbf{P}}_k^T\widehat{\mathbf{P}}_k = \mathbf{I}_k$ and $\widehat{\mathbf{P}}_{m-k}^T\widehat{\mathbf{P}}_{m-k} = \mathbf{I}_{m-k}$:

$$\|\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n\|^2 = \|\mathbf{z}_{i,k}\|^2 + \|\mathbf{z}_{i,m-k}\|^2 \checkmark$$

---

**Diagnostic Contrast: Hotelling's $T^2$ vs the $Q$ Statistic (SPE)**

| Statistic | Measures | Sensitive to |
|-----------|----------|--------------|
| $T^2_i = \sum_{j=1}^k \frac{z_{i,j}^2}{\widehat{\lambda}_j}$ | Mahalanobis distance within the principal subspace | Changes **inside** the retained subspace |
| $Q_i = \sum_{j=k+1}^m z_{i,j}^2 = \|\mathbf{e}_i\|^2$ | Squared projection error outside the principal subspace | Changes **orthogonal** to the retained subspace |

**Event Matching:**

- **Extreme operational wind loads** → **Hotelling's $T^2$**. An environmental load surge increases the amplitude of structural vibrations along the already-known principal modes (inside the retained subspace). The variance profile does not change shape; only the score magnitudes inflate. $T^2$ is calibrated to detect this elevated distance within the principal space.

- **Internal fatigue crack** → **$Q$ Statistic (SPE)**. A crack structurally re-routes load pathways, creating new vibration patterns that do not project onto the baseline principal axes. Energy appears in the residual directions outside the model subspace, driving the SPE upward even if the $T^2$ remains normal.

## **Part B: Numerical Verification — PCA Pipeline**

```
import numpy as np
import pandas as pd
from typing import Optional, Sequence, Dict, Any

class PCADiagnosticsEngine:
    def __init__(self, df: pd.DataFrame):
        self.df = df

    def compute_empirical_pca(
        self,
        columns: Optional[Sequence[str]] = None,
        show_plot: bool = True
    ) -> Dict[str, Any]:
        """
        Operationalizes the geometric de-correlation framework via PCA.
        Decomposes the unbiased empirical covariance matrix S into its orthogonal
        basis P_hat and evaluates T² and Q statistics across all possible subspace cutoffs.
        """
        import plotly.graph_objects as go
        from plotly.subplots import make_subplots

        if self.df is None:
            raise ValueError("Error: No data loaded.")

        target_cols = (
            self.df.select_dtypes(include=[np.number]).columns.tolist()
            if columns is None else list(columns)
        )
        X = self.df[target_cols].copy().dropna().values
        n, m = X.shape

        if n <= m:
            raise ValueError("Sample size n must exceed feature dimension m.")

        # 1. Center
        mu_hat = np.mean(X, axis=0)
        X_centered = X - mu_hat

        # 2. Unbiased sample covariance matrix S
        S_matrix = np.cov(X, rowvar=False, ddof=1)

        # 3. Spectral decomposition (sorted descending)
        eigenvalues, eigenvectors = np.linalg.eigh(S_matrix)
        idx = np.argsort(eigenvalues)[::-1]
        lambda_hat = np.clip(eigenvalues[idx], a_min=1e-15, a_max=None)
        P_hat = eigenvectors[:, idx]

        # 4. Variance metrics
        total_variance = np.sum(lambda_hat)
        explained_ratio = lambda_hat / total_variance
        cumulative_ratio = np.cumsum(explained_ratio)
        unexplained_ratio = 1.0 - cumulative_ratio

        # 5. Principal component scores
        Z_scores = X_centered @ P_hat

        # 6. T² and Q statistics for k in {1, ..., m-1}
        k_range = np.arange(1, m)
        mean_T2_vs_k, mean_Q_vs_k = [], []
        T2_matrix = np.zeros((n, len(k_range)))
        Q_matrix = np.zeros((n, len(k_range)))

        for idx_k, k in enumerate(k_range):
            Z_k = Z_scores[:, :k]
            lambda_k = lambda_hat[:k]
            T2_samples = np.sum((Z_k ** 2) / lambda_k, axis=1)
            T2_matrix[:, idx_k] = T2_samples
            mean_T2_vs_k.append(np.mean(T2_samples))

            Z_residual = Z_scores[:, k:]
            Q_samples = np.sum(Z_residual ** 2, axis=1)
            Q_matrix[:, idx_k] = Q_samples
            mean_Q_vs_k.append(np.mean(Q_samples))

        print(f"\n--- PCA Geometric De-correlation Layer ---")
        print(f"Feature space: {m} channels | Snapshot archive: {n} samples")
        print(f"Total System Variance (Trace[S]): {total_variance:.4f}")
        for k, ev, cv in zip(range(1, m+1), explained_ratio, cumulative_ratio):
            print(f"  PC {k}: Eigenvalue = {lambda_hat[k-1]:.4f} | "
                  f"Marginal = {ev*100:.1f}% | Cumulative = {cv*100:.1f}%")

        if show_plot:
            pc_labels = [f"PC {i+1}" for i in range(m)]
            k_labels  = [f"k={k}" for k in k_range]

            fig = make_subplots(
                rows=2, cols=3,
                horizontal_spacing=0.18, vertical_spacing=0.28,
                subplot_titles=(
                    "Feature Loading Matrix |P̂|",
                    "Component Values (Eigenvalues λ)",
                    "Information Profile (Explained Var.)",
                    "Residual Space (Unexplained Var.)",
                    "Mean Hotelling's T² vs Subspace Size k",
                    "Mean Q Statistic (SPE) vs Subspace Size k"
                )
            )

            # (1,1) Heatmap
            fig.add_trace(go.Heatmap(
                z=np.abs(P_hat), x=pc_labels, y=target_cols,
                colorscale='YlOrRd',
                colorbar=dict(title="Loading", x=-0.12, len=0.38, y=0.78,
                              yanchor="middle", xanchor="right"),
                showscale=True, showlegend=False
            ), row=1, col=1)

            # (1,2) Eigenvalues
            fig.add_trace(go.Bar(
                x=pc_labels, y=lambda_hat, name="Eigenvalue (λ_j)",
                marker=dict(color='#1f77b4', line=dict(color='black', width=0.5))
            ), row=1, col=2)
            fig.update_yaxes(title_text="Variance Magnitude", row=1, col=2)

            # (1,3) Explained Var + Cumulative
            fig.add_trace(go.Bar(
                x=pc_labels, y=explained_ratio * 100, name="Marginal Explained",
                marker=dict(color='#ff7f0e', opacity=0.75)
            ), row=1, col=3)
            fig.add_trace(go.Scatter(
                x=pc_labels, y=cumulative_ratio * 100, mode='lines+markers',
                name='Cumulative Captured',
                line=dict(color='#d62728', width=2.5, dash='dash')
            ), row=1, col=3)
            fig.update_yaxes(title_text="Captured Structure (%)", range=[-2, 105], row=1, col=3)

            # (2,1) Unexplained
            fig.add_trace(go.Bar(
                x=pc_labels, y=unexplained_ratio * 100, name="Remaining Noise",
                marker=dict(color='#2ca02c', line=dict(color='black', width=0.5))
            ), row=2, col=1)
            fig.update_yaxes(title_text="Excluded Info (%)", range=[-2, 105], row=2, col=1)

            # (2,2) T²
            fig.add_trace(go.Scatter(
                x=k_labels, y=mean_T2_vs_k, mode='lines+markers', name='Mean T²',
                line=dict(color='#9467bd', width=2.5),
                marker=dict(size=6, symbol='diamond')
            ), row=2, col=2)
            fig.update_yaxes(title_text="Average T² Metric", row=2, col=2)
            fig.update_xaxes(title_text="Truncation Cutoff (k)", row=2, col=2)

            # (2,3) Q / SPE
            fig.add_trace(go.Scatter(
                x=k_labels, y=mean_Q_vs_k, mode='lines+markers', name='Mean Q (SPE)',
                line=dict(color='#e377c2', width=2.5),
                marker=dict(size=6, symbol='square')
            ), row=2, col=3)
            fig.update_yaxes(title_text="Average Residual Energy", row=2, col=3)
            fig.update_xaxes(title_text="Truncation Cutoff (k)", row=2, col=3)

            fig.update_layout(
                title=dict(text="PCA Optimization & Feature Loading Dashboard",
                           x=0.5, y=0.97, xanchor="center"),
                template="plotly_white", showlegend=True,
                legend=dict(orientation="h", yanchor="bottom", y=1.02,
                            xanchor="center", x=0.5),
                margin=dict(t=150, b=60, l=140, r=80),
                height=750, width=1250
            )
            fig.show()

        return {
            "mean_vector": mu_hat, "covariance_matrix_S": S_matrix,
            "eigenvalues_lambda": lambda_hat, "eigenvectors_P": P_hat,
            "explained_variance_ratio": explained_ratio,
            "cumulative_variance_ratio": cumulative_ratio,
            "unexplained_variance_ratio": unexplained_ratio,
            "transformed_scores_Z": Z_scores,
            "k_values": k_range,
            "mean_T2_profile": np.array(mean_T2_vs_k),
            "mean_Q_profile": np.array(mean_Q_vs_k),
            "features": target_cols
        }


# --- Standalone test with synthetic 4-sensor data (n=3000) ---
np.random.seed(42)
n_test, m_test = 3000, 4
X_test = np.random.multivariate_normal(
    mean=[0]*m_test,
    cov=[[1.0, 0.8, 0.5, 0.1],
         [0.8, 1.0, 0.6, 0.2],
         [0.5, 0.6, 1.0, 0.3],
         [0.1, 0.2, 0.3, 1.0]],
    size=n_test
)
df_test = pd.DataFrame(X_test, columns=[f'Sensor_{i+1}' for i in range(m_test)])

engine_pca = PCADiagnosticsEngine(df_test)
pca_results = engine_pca.compute_empirical_pca(show_plot=True)
```

# **Latent Subspace Decomposition via Factor Analysis (FA)**

## **Part A: Theoretical Fundamentals**

### 1(a): Fundamental Equation of Factor Analysis

**Proof of $\mathbf{R} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}$**

Starting from the linear generative model $\mathbf{Z}_i = \boldsymbol{\Lambda}\mathbf{F}_i + \boldsymbol{\epsilon}_i$, we compute the population covariance matrix $\mathbf{R} = \mathbb{E}[\mathbf{Z}_i\mathbf{Z}_i^T]$ directly:

$$\mathbf{R} = \mathbb{E}[(\boldsymbol{\Lambda}\mathbf{F}_i + \boldsymbol{\epsilon}_i)(\boldsymbol{\Lambda}\mathbf{F}_i + \boldsymbol{\epsilon}_i)^T]$$

Expanding the outer product:

$$= \mathbb{E}[\boldsymbol{\Lambda}\mathbf{F}_i\mathbf{F}_i^T\boldsymbol{\Lambda}^T] + \mathbb{E}[\boldsymbol{\Lambda}\mathbf{F}_i\boldsymbol{\epsilon}_i^T] + \mathbb{E}[\boldsymbol{\epsilon}_i\mathbf{F}_i^T\boldsymbol{\Lambda}^T] + \mathbb{E}[\boldsymbol{\epsilon}_i\boldsymbol{\epsilon}_i^T]$$

Applying the model assumptions:
- $\mathbb{E}[\mathbf{F}_i\mathbf{F}_i^T] = \mathbf{I}_{k \times k}$ (orthonormal factors)
- $\mathbb{E}[\boldsymbol{\epsilon}_i\mathbf{F}_i^T] = \mathbf{0}$ (independence of factors and errors)
- $\mathbb{E}[\boldsymbol{\epsilon}_i\boldsymbol{\epsilon}_i^T] = \boldsymbol{\Psi} = \text{diag}(\varphi_1^2, \dots, \varphi_m^2)$

$$\mathbf{R} = \boldsymbol{\Lambda}\mathbf{I}\boldsymbol{\Lambda}^T + \mathbf{0} + \mathbf{0} + \boldsymbol{\Psi} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi} \checkmark$$

---

**Communality and Uniqueness — Physical Interpretations**

Focusing on the $j$-th diagonal entry of the fundamental equation:

$$R_{jj} = 1 = \underbrace{\sum_{r=1}^k \lambda_{j,r}^2}_{h_j^2} + \underbrace{\varphi_j^2}_{\text{uniqueness}}$$

- **Communality $h_j^2 = \sum_{r=1}^k \lambda_{j,r}^2$:** The fraction of sensor $j$'s total variance that is *shared* with the common latent physical factors. In an SHM system, high communality indicates the sensor is responding to genuine structural dynamics.

- **Uniqueness $\varphi_j^2 = 1 - h_j^2$:** The fraction of sensor $j$'s variance that is completely isolated — not shared with any other channel. In asset diagnostics, high uniqueness flags potential **instrument malfunction, calibration drift, or localized electrical noise** on that particular sensor.

### 1(b): Varimax Rotation — Conceptual Contrast

**Raw PCA loading vector vs Varimax-rotated FA loading vector:**

A raw PCA eigenvector is a mathematical solution to a constrained global optimization — it represents the direction in sensor space that captures the maximum possible total variance from the data cloud. Consequently, each eigenvector tends to spread its loading energy across *all* sensors simultaneously, producing dense loading vectors that are difficult to assign physical meaning to.

A **Varimax-rotated** factor loading vector maximizes the variance of the squared loadings within each column of $\boldsymbol{\Lambda}$. This drives loading patterns toward "simple structure": most entries become either near-zero or near-one, cleanly grouping sensors into distinct physical subsets.

---

**Why orthogonal rotations preserve $h_j^2$ and $\mathbf{R} \approx \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}$:**

An orthogonal rotation replaces $\boldsymbol{\Lambda}$ with $\boldsymbol{\Lambda}^* = \boldsymbol{\Lambda}\mathbf{T}$, where $\mathbf{T}^T\mathbf{T} = \mathbf{I}$. The communality for sensor $j$ under the rotation is:

$$\sum_{r=1}^k (\lambda_{j,r}^*)^2 = \left\|\boldsymbol{\Lambda}_{j,\cdot}\mathbf{T}\right\|^2 = \boldsymbol{\Lambda}_{j,\cdot}\mathbf{T}\mathbf{T}^T\boldsymbol{\Lambda}_{j,\cdot}^T = \boldsymbol{\Lambda}_{j,\cdot}\boldsymbol{\Lambda}_{j,\cdot}^T = h_j^2$$

The row norm is unchanged by orthogonal multiplication. Similarly:

$$\boldsymbol{\Lambda}^*(\boldsymbol{\Lambda}^*)^T = (\boldsymbol{\Lambda}\mathbf{T})(\boldsymbol{\Lambda}\mathbf{T})^T = \boldsymbol{\Lambda}\mathbf{T}\mathbf{T}^T\boldsymbol{\Lambda}^T = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T$$

Therefore $\mathbf{R} \approx \boldsymbol{\Lambda}^*(\boldsymbol{\Lambda}^*)^T + \boldsymbol{\Psi}$ is identical to the un-rotated approximation. Rotation is a pure re-labelling of the latent coordinate axes — it changes interpretability without changing fit.

### 2: Thomson's Score Matrix Derivation

**Step 1: Joint Covariance Structure of $\mathbf{Y}_i = [\mathbf{Z}_i^T, \mathbf{F}_i^T]^T$**

Computing each block of $\boldsymbol{\Sigma}_{YY} = \mathbb{E}[\mathbf{Y}_i\mathbf{Y}_i^T]$:

- **Block (1,1):** $\mathbb{E}[\mathbf{Z}_i\mathbf{Z}_i^T] = \mathbf{R}$
- **Block (2,2):** $\mathbb{E}[\mathbf{F}_i\mathbf{F}_i^T] = \mathbf{I}_{k \times k}$
- **Block (1,2):** $\mathbb{E}[\mathbf{Z}_i\mathbf{F}_i^T] = \mathbb{E}[(\boldsymbol{\Lambda}\mathbf{F}_i + \boldsymbol{\epsilon}_i)\mathbf{F}_i^T] = \boldsymbol{\Lambda}\underbrace{\mathbb{E}[\mathbf{F}_i\mathbf{F}_i^T]}_{\mathbf{I}} + \underbrace{\mathbb{E}[\boldsymbol{\epsilon}_i\mathbf{F}_i^T]}_{\mathbf{0}} = \boldsymbol{\Lambda}$
- **Block (2,1):** $\boldsymbol{\Lambda}^T$

Therefore:

$$\boldsymbol{\Sigma}_{YY} = \begin{bmatrix} \mathbf{R} & \boldsymbol{\Lambda} \\ \boldsymbol{\Lambda}^T & \mathbf{I}_{k \times k} \end{bmatrix} \checkmark$$

---

**Step 2: Thomson's Regression Score Derivation**

Under the Multivariate Normal assumption with zero means ($\boldsymbol{\mu}_1 = \mathbf{0}$, $\boldsymbol{\mu}_2 = \mathbf{0}$), the conditional mean formula gives:

$$\mathbf{f}_i = \mathbb{E}[\mathbf{F}_i | \mathbf{z}_i] = \boldsymbol{\mu}_2 + \boldsymbol{\Sigma}_{21}\boldsymbol{\Sigma}_{11}^{-1}(\mathbf{z}_i - \boldsymbol{\mu}_1)$$

Substituting $\boldsymbol{\Sigma}_{21} = \boldsymbol{\Lambda}^T$ and $\boldsymbol{\Sigma}_{11} = \mathbf{R}$:

$$\mathbf{f}_i = \boldsymbol{\Lambda}^T\mathbf{R}^{-1}\mathbf{z}_i \checkmark$$

This is Thomson's MMSE estimator — it weights the normalized observation $\mathbf{z}_i$ by the correlation-weighted loading matrix to recover the best linear unbiased estimate of the hidden factor realization.

## **Part B: FA Parametric Engine & Dashboard**

```
import numpy as np
import pandas as pd
from typing import Optional, Sequence, Dict, Any
from sklearn.decomposition import FactorAnalysis
import plotly.graph_objects as go
from plotly.subplots import make_subplots


class FADiagnosticsEngine:
    def __init__(self, df: pd.DataFrame):
        self.df = df

    def compute_empirical_fa(
        self,
        k: int,
        columns: Optional[Sequence[str]] = None,
        show_plot: bool = True
    ) -> Dict[str, Any]:
        """
        Operationalizes the Factor Analysis latent subspace framework with Varimax rotation.
        Renders a clean 2×2 diagnostic dashboard.
        """
        if self.df is None:
            raise ValueError("Error: No data loaded.")

        target_cols = (
            self.df.select_dtypes(include=[np.number]).columns.tolist()
            if columns is None else list(columns)
        )
        X = self.df[target_cols].copy().dropna().values
        n, m = X.shape

        # --- Dimensionality Constraint Layer ---
        if n <= m:
            raise ValueError(f"Snapshot count (n={n}) must exceed channel count (m={m}).")
        if k >= m:
            raise ValueError(f"Latent factor count (k={k}) must be strictly less than feature count (m={m}).")

        # --- Z-Score Mapping Layer ---
        mu_hat = np.mean(X, axis=0)
        std_hat = np.std(X, axis=0, ddof=1)
        std_hat[std_hat == 0] = 1e-15                 # numerical guardrail
        Z = (X - mu_hat) / std_hat
        R_matrix = np.corrcoef(X, rowvar=False)

        # --- Information Decomposition Layer ---
        fa = FactorAnalysis(n_components=k, rotation='varimax', random_state=42)
        fa.fit(Z)
        lambda_matrix = fa.components_.T             # shape (m, k)
        uniqueness    = fa.noise_variance_            # φ²
        communality   = np.sum(lambda_matrix**2, axis=1)  # h²

        # --- Latent Score Projection (Thomson's Method) ---
        F_scores = fa.transform(Z)                   # shape (n, k)

        # --- Performance Diagnostic Telemetry ---
        factor_variances = np.var(F_scores, axis=0, ddof=1)
        avg_communality  = np.mean(communality) * 100
        avg_uniqueness   = np.mean(uniqueness)  * 100

        print("\n" + "=" * 60)
        print("  FA Latent Subspace Diagnostics Report")
        print("=" * 60)
        print(f"  Sensor Space  : {m} channels | Archive : {n} snapshots")
        print(f"  Latent Factors: k = {k}  (Varimax rotation)")
        print(f"  Avg System Communality (h²) : {avg_communality:.2f}%")
        print(f"  Avg System Uniqueness  (φ²) : {avg_uniqueness:.2f}%")
        for j, (h2, phi2) in enumerate(zip(communality, uniqueness)):
            print(f"  {target_cols[j]:12s}  h² = {h2*100:.1f}%   φ² = {phi2*100:.1f}%")
        print("=" * 60)

        if show_plot:
            sensor_labels = target_cols
            factor_labels = [f"Factor {j+1}" for j in range(k)]

            fig = make_subplots(
                rows=2, cols=2,
                horizontal_spacing=0.24, vertical_spacing=0.28,
                subplot_titles=(
                    "Structural Loadings Matrix |λ_(j,r)|",
                    "Variance Partitioning (Communality vs Uniqueness)",
                    "Sensor Uniqueness Noise Floor (φ²)",
                    "Latent Factor Scores Empirical Variance"
                )
            )

            # (1,1) Heatmap
            fig.add_trace(go.Heatmap(
                z=np.abs(lambda_matrix), x=factor_labels, y=sensor_labels,
                colorscale='YlOrRd',
                colorbar=dict(title="Sensitivity", x=-0.15, len=0.38, y=0.78,
                              yanchor="middle", xanchor="right"),
                showscale=True, showlegend=False
            ), row=1, col=1)
            fig.update_xaxes(title_text="Latent Structures", row=1, col=1)

            # (1,2) Stacked horizontal bar
            fig.add_trace(go.Bar(
                y=sensor_labels, x=communality * 100,
                name="Communality (h² – Shared Structure)",
                orientation='h', marker=dict(color='#1f77b4')
            ), row=1, col=2)
            fig.add_trace(go.Bar(
                y=sensor_labels, x=uniqueness * 100,
                name="Uniqueness (φ² – Channel Noise)",
                orientation='h', marker=dict(color='#ff7f0e')
            ), row=1, col=2)
            fig.update_layout(barmode='stack')
            fig.update_xaxes(title_text="Variance Allocation (%)", range=[0, 100], row=1, col=2)

            # (2,1) Uniqueness line
            fig.add_trace(go.Scatter(
                x=sensor_labels, y=uniqueness, mode='lines+markers',
                name="Uniqueness Profile (φ²)",
                line=dict(color='#d62728', width=2, dash='dot'),
                marker=dict(size=8, symbol='x')
            ), row=2, col=1)
            fig.update_yaxes(range=[-0.05, 1.05], row=2, col=1)
            fig.update_xaxes(title_text="Monitored Channels", tickangle=25, row=2, col=1)

            # (2,2) Factor variance bars
            fig.add_trace(go.Bar(
                x=factor_labels, y=factor_variances,
                name="Factor Empirical Variance",
                marker=dict(color='#2ca02c', line=dict(color='black', width=0.5))
            ), row=2, col=2)
            fig.update_yaxes(title_text="Variance Level", row=2, col=2)
            fig.update_xaxes(title_text="Latent Vectors", row=2, col=2)

            fig.update_layout(
                title=dict(text="Factor Analysis (FA) Latent Subspace Diagnostics Dashboard",
                           x=0.5, y=0.97, xanchor="center"),
                template="plotly_white", showlegend=True,
                legend=dict(orientation="h", yanchor="bottom", y=1.02,
                            xanchor="center", x=0.5),
                margin=dict(t=150, b=60, l=140, r=80),
                height=750, width=1250
            )
            fig.show()

        return {
            "mean_vector_mu": mu_hat, "std_vector_D": std_hat,
            "correlation_matrix_R": R_matrix,
            "factor_loadings_lambda": lambda_matrix,
            "uniqueness_psi": uniqueness, "communality_h2": communality,
            "latent_factor_scores_F": F_scores,
            "factor_variances": factor_variances,
            "sensors": target_cols
        }
```

```
# =====================================================================
# COMBINED SUBSPACE ENGINES — PCA & FA PIPELINE
# =====================================================================

import numpy as np
import pandas as pd
from typing import Optional, Sequence, Dict, Any
from sklearn.decomposition import FactorAnalysis
import plotly.graph_objects as go
from plotly.subplots import make_subplots


class DualSubspaceDiagnosticsEngine:
    def __init__(self, df: pd.DataFrame):
        self.df = df

    # ------------------------------------------------------------------
    def compute_empirical_pca(
        self,
        columns: Optional[Sequence[str]] = None,
        show_plot: bool = True
    ) -> Dict[str, Any]:
        """
        PCA pipeline with 2×3 optimization dashboard.
        Panels: loading heatmap | eigenvalues | explained var
                unexplained var | mean T²      | mean Q (SPE)
        """
        target_cols = (
            self.df.select_dtypes(include=[np.number]).columns.tolist()
            if columns is None else list(columns)
        )
        X = self.df[target_cols].copy().dropna().values
        n, m = X.shape
        if n <= m:
            raise ValueError("n must exceed m.")

        mu_hat     = np.mean(X, axis=0)
        X_centered = X - mu_hat
        S_matrix   = np.cov(X, rowvar=False, ddof=1)

        eigenvalues, eigenvectors = np.linalg.eigh(S_matrix)
        idx        = np.argsort(eigenvalues)[::-1]
        lambda_hat = np.clip(eigenvalues[idx], 1e-15, None)
        P_hat      = eigenvectors[:, idx]

        total_var         = np.sum(lambda_hat)
        explained_ratio   = lambda_hat / total_var
        cumulative_ratio  = np.cumsum(explained_ratio)
        unexplained_ratio = 1.0 - cumulative_ratio

        Z_scores = X_centered @ P_hat
        k_range  = np.arange(1, m)
        mean_T2, mean_Q = [], []
        T2_mat = np.zeros((n, len(k_range)))
        Q_mat  = np.zeros((n, len(k_range)))

        for i, k in enumerate(k_range):
            T2s = np.sum((Z_scores[:, :k] ** 2) / lambda_hat[:k], axis=1)
            Qs  = np.sum(Z_scores[:, k:] ** 2, axis=1)
            T2_mat[:, i] = T2s;  mean_T2.append(np.mean(T2s))
            Q_mat[:, i]  = Qs;   mean_Q.append(np.mean(Qs))

        print(f"\n--- PCA Layer | {m} features | {n} samples ---")
        print(f"Total Variance (Trace[S]): {total_var:.4f}")
        for j in range(m):
            print(f"  PC {j+1}: λ={lambda_hat[j]:.4f} | "
                  f"{explained_ratio[j]*100:.1f}% | cum {cumulative_ratio[j]*100:.1f}%")

        if show_plot:
            pc_labels = [f"PC {i+1}" for i in range(m)]
            k_labels  = [f"k={k}"   for k in k_range]

            fig = make_subplots(
                rows=2, cols=3,
                horizontal_spacing=0.18, vertical_spacing=0.28,
                subplot_titles=(
                    "Feature Loading Matrix |P̂|",
                    "Component Values (Eigenvalues λ)",
                    "Information Profile (Explained Var.)",
                    "Residual Space (Unexplained Var.)",
                    "Mean Hotelling's T² vs Subspace Size k",
                    "Mean Q Statistic (SPE) vs Subspace Size k"
                )
            )
            fig.add_trace(go.Heatmap(
                z=np.abs(P_hat), x=pc_labels, y=target_cols,
                colorscale='YlOrRd',
                colorbar=dict(title="Loading", x=-0.12, len=0.38, y=0.78,
                              yanchor="middle", xanchor="right"),
                showscale=True, showlegend=False
            ), row=1, col=1)
            fig.add_trace(go.Bar(x=pc_labels, y=lambda_hat, name="Eigenvalue (λ_j)",
                marker=dict(color='#1f77b4', line=dict(color='black', width=0.5))), row=1, col=2)
            fig.update_yaxes(title_text="Variance Magnitude", row=1, col=2)
            fig.add_trace(go.Bar(x=pc_labels, y=explained_ratio*100, name="Marginal Explained",
                marker=dict(color='#ff7f0e', opacity=0.75)), row=1, col=3)
            fig.add_trace(go.Scatter(x=pc_labels, y=cumulative_ratio*100, mode='lines+markers',
                name='Cumulative Captured', line=dict(color='#d62728', width=2.5, dash='dash')), row=1, col=3)
            fig.update_yaxes(title_text="Captured (%)", range=[-2, 105], row=1, col=3)
            fig.add_trace(go.Bar(x=pc_labels, y=unexplained_ratio*100, name="Remaining Noise",
                marker=dict(color='#2ca02c', line=dict(color='black', width=0.5))), row=2, col=1)
            fig.update_yaxes(title_text="Excluded Info (%)", range=[-2, 105], row=2, col=1)
            fig.add_trace(go.Scatter(x=k_labels, y=mean_T2, mode='lines+markers', name='Mean T²',
                line=dict(color='#9467bd', width=2.5), marker=dict(size=6, symbol='diamond')), row=2, col=2)
            fig.update_yaxes(title_text="Average T² Metric", row=2, col=2)
            fig.update_xaxes(title_text="Truncation Cutoff (k)", row=2, col=2)
            fig.add_trace(go.Scatter(x=k_labels, y=mean_Q, mode='lines+markers', name='Mean Q (SPE)',
                line=dict(color='#e377c2', width=2.5), marker=dict(size=6, symbol='square')), row=2, col=3)
            fig.update_yaxes(title_text="Average Residual Energy", row=2, col=3)
            fig.update_xaxes(title_text="Truncation Cutoff (k)", row=2, col=3)
            fig.update_layout(
                title=dict(text="Principal Component Analysis (PCA) Optimization & Feature Loading Dashboard",
                           x=0.5, y=0.97, xanchor="center"),
                template="plotly_white", showlegend=True,
                legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="center", x=0.5),
                margin=dict(t=150, b=60, l=140, r=80), height=750, width=1250
            )
            fig.show()

        return {
            "mean_vector": mu_hat, "covariance_matrix_S": S_matrix,
            "eigenvalues_lambda": lambda_hat, "eigenvectors_P": P_hat,
            "explained_variance_ratio": explained_ratio,
            "cumulative_variance_ratio": cumulative_ratio,
            "unexplained_variance_ratio": unexplained_ratio,
            "transformed_scores_Z": Z_scores, "k_values": k_range,
            "mean_T2_profile": np.array(mean_T2), "mean_Q_profile": np.array(mean_Q),
            "features": target_cols
        }

    # ------------------------------------------------------------------
    def compute_empirical_fa(
        self,
        k: int,
        columns: Optional[Sequence[str]] = None,
        show_plot: bool = True
    ) -> Dict[str, Any]:
        """
        FA pipeline with Varimax rotation and 2×2 diagnostic dashboard.
        """
        target_cols = (
            self.df.select_dtypes(include=[np.number]).columns.tolist()
            if columns is None else list(columns)
        )
        X = self.df[target_cols].copy().dropna().values
        n, m = X.shape
        if n <= m:
            raise ValueError(f"n={n} must exceed m={m}.")
        if k >= m:
            raise ValueError(f"k={k} must be < m={m}.")

        mu_hat  = np.mean(X, axis=0)
        std_hat = np.std(X, axis=0, ddof=1)
        std_hat[std_hat == 0] = 1e-15
        Z        = (X - mu_hat) / std_hat
        R_matrix = np.corrcoef(X, rowvar=False)

        fa = FactorAnalysis(n_components=k, rotation='varimax', random_state=42)
        fa.fit(Z)
        lambda_matrix = fa.components_.T
        uniqueness    = fa.noise_variance_
        communality   = np.sum(lambda_matrix**2, axis=1)
        F_scores      = fa.transform(Z)
        factor_variances = np.var(F_scores, axis=0, ddof=1)

        print(f"\n--- FA Layer | {m} sensors → k={k} factors (Varimax) ---")
        print(f"Avg Communality: {np.mean(communality)*100:.2f}%  |  "
              f"Avg Uniqueness: {np.mean(uniqueness)*100:.2f}%")
        for j, (h2, phi2) in enumerate(zip(communality, uniqueness)):
            print(f"  {target_cols[j]:12s}  h²={h2*100:.1f}%  φ²={phi2*100:.1f}%")

        if show_plot:
            sensor_labels = target_cols
            factor_labels = [f"Factor {j+1}" for j in range(k)]

            fig = make_subplots(
                rows=2, cols=2,
                horizontal_spacing=0.24, vertical_spacing=0.28,
                subplot_titles=(
                    "Structural Loadings Matrix |λ_(j,r)|",
                    "Variance Partitioning (Communality vs Uniqueness)",
                    "Sensor Uniqueness Noise Floor (φ²)",
                    "Latent Factor Scores Empirical Variance"
                )
            )
            fig.add_trace(go.Heatmap(
                z=np.abs(lambda_matrix), x=factor_labels, y=sensor_labels,
                colorscale='YlOrRd',
                colorbar=dict(title="Sensitivity", x=-0.15, len=0.38, y=0.78,
                              yanchor="middle", xanchor="right"),
                showscale=True, showlegend=False
            ), row=1, col=1)
            fig.update_xaxes(title_text="Latent Structures", row=1, col=1)
            fig.add_trace(go.Bar(y=sensor_labels, x=communality*100,
                name="Communality (h² – Shared Structure)",
                orientation='h', marker=dict(color='#1f77b4')), row=1, col=2)
            fig.add_trace(go.Bar(y=sensor_labels, x=uniqueness*100,
                name="Uniqueness (φ² – Channel Noise)",
                orientation='h', marker=dict(color='#ff7f0e')), row=1, col=2)
            fig.update_layout(barmode='stack')
            fig.update_xaxes(title_text="Variance Allocation (%)", range=[0, 100], row=1, col=2)
            fig.add_trace(go.Scatter(x=sensor_labels, y=uniqueness, mode='lines+markers',
                name="Uniqueness Profile (φ²)",
                line=dict(color='#d62728', width=2, dash='dot'),
                marker=dict(size=8, symbol='x')), row=2, col=1)
            fig.update_yaxes(range=[-0.05, 1.05], row=2, col=1)
            fig.update_xaxes(title_text="Monitored Channels", tickangle=25, row=2, col=1)
            fig.add_trace(go.Bar(x=factor_labels, y=factor_variances,
                name="Factor Empirical Variance",
                marker=dict(color='#2ca02c', line=dict(color='black', width=0.5))), row=2, col=2)
            fig.update_yaxes(title_text="Variance Level", row=2, col=2)
            fig.update_xaxes(title_text="Latent Vectors", row=2, col=2)
            fig.update_layout(
                title=dict(text="Factor Analysis (FA) Latent Subspace Diagnostics Dashboard",
                           x=0.5, y=0.97, xanchor="center"),
                template="plotly_white", showlegend=True,
                legend=dict(orientation="h", yanchor="bottom", y=1.02,
                            xanchor="center", x=0.5,
                            traceorder="normal", itemwidth=40),
                margin=dict(t=150, b=60, l=140, r=80),
                height=750, width=1250
            )
            fig.show()

        return {
            "mean_vector_mu": mu_hat, "std_vector_D": std_hat,
            "correlation_matrix_R": R_matrix,
            "factor_loadings_lambda": lambda_matrix,
            "uniqueness_psi": uniqueness, "communality_h2": communality,
            "latent_factor_scores_F": F_scores,
            "factor_variances": factor_variances,
            "sensors": target_cols
        }


# =====================================================================
# DATA GENERATOR & PIPELINE EXECUTION
# =====================================================================
np.random.seed(42)
n_samples = 2500

f1 = np.random.normal(0, 1, n_samples)
f2 = np.random.normal(0, 1, n_samples)

s1 = 0.85 * f1 + 0.10 * f2 + np.random.normal(0, 0.30, n_samples)
s2 = 0.80 * f1 + 0.15 * f2 + np.random.normal(0, 0.35, n_samples)
s3 = 0.12 * f1 + 0.90 * f2 + np.random.normal(0, 0.25, n_samples)
s4 = 0.02 * f1 + 0.05 * f2 + np.random.normal(0, 1.40, n_samples)

df_asset = pd.DataFrame(
    data=np.vstack([s1, s2, s3, s4]).T,
    columns=['Sensor_1', 'Sensor_2', 'Sensor_3', 'Sensor_4']
)

engine = DualSubspaceDiagnosticsEngine(df_asset)

# Run PCA (2×3 Dashboard)
pca_metrics = engine.compute_empirical_pca(show_plot=True)

# Run FA (2×2 Dashboard, k=2 common factors)
fa_metrics = engine.compute_empirical_fa(k=2, show_plot=True)
```

## **Part B: Analysis — Answers**

### Question 1: The Total Variance Illusion

**Part 1 — Physical nature of Sensor_4 (φ² ≈ 98%):**

A uniqueness value approaching $100\%$ indicates that `Sensor_4` shares virtually zero covariance with the rest of the monitoring array. It is completely decoupled from the real physical processes ($f_1$, $f_2$) driving the structural asset. Its entire data profile consists of isolated white noise — the result of a severe electrical fault or instrumentation failure — with no genuine structural signal content whatsoever.

**Part 2 — How Sensor_4 noise "tricks" the PCA engine:**

PCA maximizes *total* global variance without distinguishing its origin. Because `Sensor_4` carries a massive noise-driven variance ($\sigma^2 \approx 2.0$), far exceeding the shared structural variance of `Sensor_1`–`Sensor_3` (each $\approx 0.8$–$1.0$), the eigen-decomposition of $\mathbf{S}$ places the direction aligned with `Sensor_4`'s random scatter as the *largest* eigenvalue direction ($\lambda_1 \approx 2.0$). `PC 1` becomes almost entirely dedicated to tracking this single faulty sensor, while the genuine correlated structural signals of the healthy sensors are pushed to lower principal components with reduced eigenvalues.

**Part 3 — Engineering risk of a PCA-only monitoring framework:**

If a monitoring system relies solely on PCA, a single blown or short-circuited instrument will silently hijack the most critical surveillance axis ($PC\ 1$), inflating $\lambda_1$ and rendering the primary control chart insensitive to true structural anomalies. The Hotelling's $T^2$ chart trained on contaminated principal axes will generate high false-alarm rates for irrelevant electrical noise while simultaneously missing genuine structural damage events. This represents a critical safety failure in an operational SHM context.

### Question 2: Decoupling Structural Loading via Varimax Rotation

**Part 1 — PCA strict hierarchy vs. Varimax "simple structure":**

Traditional PCA is bound to a strict mathematical hierarchy: each successive eigenvector must (a) be orthogonal to all previous ones and (b) maximize the remaining unexplained variance across *all* sensors simultaneously. This global variance-maximization objective forces every principal axis to receive non-zero projections from multiple sensors, producing dense loading vectors where each sensor partially "bleeds" into multiple components.

**Varimax rotation** abandons the hierarchy constraint after the initial decomposition. It applies an orthogonal rotation $\mathbf{T}$ to the loading matrix to maximize $\sum_r \sum_j \left(\frac{\lambda_{j,r}^2}{h_j^2}\right)^2 - \left(\sum_j \frac{\lambda_{j,r}^2}{h_j^2}\right)^2$, which drives each loading column toward having a few values near $\pm 1$ and the rest near $0$. The result is "simple structure": each latent factor activates a *distinct cluster* of sensors rather than a messy mixture of all.

**Part 2 — Operational troubleshooting advantage:**

From a plant operator's perspective, the rotated FA loading heatmap offers immediate clarity. When `Factor 1` maps almost exclusively to `Sensor_1` and `Sensor_2`, any alert on `Factor 1` instantly narrows the root cause to the structural subsystem those two sensors jointly monitor — no decomposition algebra required. In contrast, the PCA eigenvector matrix shows mixed, similar-magnitude loadings across multiple sensors for each component. Diagnosing which physical subsystem triggered an alarm requires solving a weighted back-projection, which is impractical in real-time operations under pressure.

### Question 3: Determining Subspace Truncation via $T^2$ and $Q$ Profiles

**Part 1 — $Q$ Statistic curve behavior:**

- **$k = 1 \to k = 2$:** The Mean $Q$ Statistic drops sharply. Retaining only `PC 1` (which tracks `Sensor_4`'s noise) leaves the structured shared variance of `Sensor_1`–`Sensor_3` entirely in the residual space, producing a very large average SPE. Adding `PC 2` captures the dominant structural mode of these three sensors, eliminating the bulk of the residual energy and causing a steep descent in the $Q$ curve.

- **$k = 2 \to k = 3$:** The curve flattens dramatically into a clear "elbow." The additional component `PC 3` absorbs only minor residual scatter from the already-explained sensors. The marginal reduction in residual energy is small and nearly linear, signalling that no new structured physical dimension exists beyond $k = 2$.

**Part 2 — Why $k = 2$ identifies the true physical dimensionality:**

The sharp drop followed by the flat elbow is the hallmark of dimensionality identification. Once $k = 2$ components account for (a) the noise-dominated `Sensor_4` direction and (b) the shared structural variance of `Sensor_1`–`Sensor_3`, all meaningful patterned structure is exhausted. The $Q$ residual beyond $k = 2$ consists only of baseline measurement noise, which decays slowly and uniformly regardless of additional dimensions. The "elbow" position at $k = 2$ therefore reveals the true intrinsic hidden dimensionality of the asset.

**Part 3 — Consequence of over-truncating at $k = 3$:**

`PC 3` absorbs the residual scatter of `Sensor_3` after its primary structural contribution is already captured in `PC 2`. Forcing this residual into the "clean" principal subspace at $k = 3$ means your Hotelling's $T^2$ monitoring chart must now account for minor, benign fluctuations in `Sensor_3`'s measurement noise as part of the structural baseline. This *artificially expands* the control limits and generates excessive false alarms whenever `Sensor_3` experiences any minor ambient variation, degrading the sensitivity of the monitoring layer to genuine structural anomalies.

### Question 4: Operational Trade-offs — PCA vs FA Strategy

**The PCA Strategy: $T^2$ + $Q$ Joint Monitoring**

The PCA approach monitors two complementary spaces: the principal subspace via Hotelling's $T^2$ (detecting shifts *within* the modelled structure) and the residual space via the $Q$ statistic (detecting novel directions of variation *outside* the model). Together they form a complete partition of the sensor space.

*Vulnerability:* As demonstrated in this dataset, a single high-noise sensor like `Sensor_4` distorts the orientation of the principal axes. If `Sensor_4` loses calibration and its noise floor escalates further, the principal axes shift, corrupting the $T^2$ control limits for all other channels simultaneously. **A single-channel failure propagates globally** through the PCA subspace.

---

**The FA Strategy: Rotated Latent Factor Score Monitoring**

The FA framework explicitly decomposes each sensor's variance into shared structural content ($h_j^2$) and isolated noise ($\varphi_j^2$). The latent factor scores track only the *shared* physical dynamics — the noise floor of each sensor is algebraically excluded.

*Robustness:* If `Sensor_4` experiences an electrical short and its noise level doubles, this perturbation is absorbed into $\varphi_4^2 \to 1$. The uniqueness layer isolates the fault, and the latent factor scores `Factor 1` and `Factor 2` — which depend on `Sensor_4` with near-zero loading — remain clean and unaffected.

**Conclusion: The FA strategy is significantly more robust** against single-sensor calibration loss or electrical faults. The uniqueness noise floor ($\varphi^2$) acts as a per-channel fault containment buffer, preventing individual instrument failures from contaminating the global structural health model. In a fleet-scale predictive maintenance deployment, this translates directly into fewer erroneous maintenance dispatches and higher confidence in genuine damage alerts.
