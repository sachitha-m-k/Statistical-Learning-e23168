# Q. Gaussian Mixture Clustering as Conditional Updating

Consider a dataset
$$
x_1,x_2,\dots,x_n\in\mathbb R^d.
$$
We wish to cluster these observations into $K$ groups. Instead of assigning each point deterministically to a cluster at the beginning, we introduce a latent random variable
$$
C_i\in{1,\dots,K},
$$
where $C_i=k$ means that the observation $x_i$ belongs to cluster $k$.
Let the prior probability of cluster membership be
$$
P(C_i=k)=\phi_k,
$$
where
$$
\phi_k\ge 0,
\qquad
\sum_{k=1}^K \phi_k=1.
$$

Conditional on $C_i=k$, assume that the observation $X_i$ is generated from a multivariate Gaussian distribution:
$$
X_i\mid C_i=k
\sim
\mathscr N(\mu_k,\Sigma_k),
$$
where
$$
\mu_k\in\mathbb R^d,
\qquad
\Sigma_k\in\mathbb R^{d\times d}
$$
are the mean vector and covariance matrix of cluster $k$.

The model parameters
$$
\phi_k,\mu_k,\Sigma_k,
\qquad k=1,\dots,K,
$$
are assumed to be fixed but unknown.

## Answer

### 1. Marginal density

By definition of a joint distribution over $(X_i, C_i)$, the marginal density of $X_i$ is obtained by summing the joint density over every possible value of the discrete latent label $C_i$ (law of total probability):

$$
p(x_i)=\sum_{k=1}^{K}p(x_i, C_i=k)=\sum_{k=1}^{K}P(C_i=k)\,p(x_i\mid C_i=k)=\sum_{k=1}^K \phi_k\,\mathscr N(x_i\mid\mu_k,\Sigma_k).
$$

This is called a **Gaussian mixture density** because it is a convex combination (weights $\phi_k\ge0$, $\sum_k\phi_k=1$) of $K$ individual Gaussian densities — the overall shape is literally a weighted "mixture" of several Gaussian bumps rather than a single one.

### 2. Posterior cluster probability

Bayes' rule applied to the discrete latent variable $C_i$ gives

$$
P(C_i=k\mid X_i=x_i)=\frac{P(X_i=x_i\mid C_i=k)\,P(C_i=k)}{\sum_{j=1}^{K}P(X_i=x_i\mid C_i=j)\,P(C_i=j)}.
$$

Substituting the Gaussian conditional density for the numerator/denominator terms and $\phi_k$ for the prior class probability:

$$
P(C_i=k\mid X_i=x_i)=\frac{\phi_k\,\mathscr N(x_i\mid\mu_k,\Sigma_k)}{\sum_{j=1}^K \phi_j\,\mathscr N(x_i\mid\mu_j,\Sigma_j)}=:\gamma_{ik}.
$$

$\gamma_{ik}$ is legitimately called a **posterior** probability of cluster membership because it is exactly the conditional probability $P(C_i=k\mid X_i=x_i)$ — it starts from the prior belief $\phi_k$ about cluster membership and updates that belief in light of the observed data $x_i$, via Bayes' theorem, just as any posterior does.

### 3. One-hot encoding and its conditional expectation

Since $Z_{ik}$ only takes values in $\{0,1\}$, its conditional expectation given $X_i=x_i$ is, by the definition of expectation for a binary random variable,

$$
\mathbb E[Z_{ik}\mid X_i=x_i] = 1\cdot P(Z_{ik}=1\mid X_i=x_i) + 0\cdot P(Z_{ik}=0\mid X_i=x_i) = P(Z_{ik}=1\mid X_i=x_i).
$$

But $Z_{ik}=1$ exactly when $C_i=k$, so $P(Z_{ik}=1\mid X_i=x_i)=P(C_i=k\mid X_i=x_i)=\gamma_{ik}$, giving

$$
\mathbb E[Z_{ik}\mid X_i=x_i]=\gamma_{ik}.
$$

Stacking this over $k=1,\dots,K$ (conditional expectation applies entrywise to a random vector),

$$
\mathbb E[Z_i\mid X_i=x_i]=\begin{bmatrix}\gamma_{i1}\\ \vdots \\ \gamma_{iK}\end{bmatrix}.
$$

So the vector of soft cluster-membership probabilities used in GMM clustering is nothing more than the conditional expectation of the one-hot cluster indicator given the data — clustering here is literally an exercise in computing $\mathbb E[Z_i\mid X_i=x_i]$.

### 4. Soft vs. hard clustering

Soft clustering keeps the *entire* probability vector $\mathbb E[Z_i\mid X_i=x_i]=(\gamma_{i1},\dots,\gamma_{iK})$ — it expresses graded, partial membership in every cluster simultaneously (e.g. a point could be 60% cluster 1, 35% cluster 2, 5% cluster 3), which is useful when clusters overlap and a point genuinely sits in an ambiguous region. Hard clustering collapses this vector down to a single deterministic label $\widehat C_i=\arg\max_k \gamma_{ik}$, discarding the uncertainty information and forcing every point into exactly one cluster. Hard assignment is simpler to report and act on, but it throws away exactly the information (how confident the assignment is) that soft clustering preserves.

### 5. Conditional expectation of the observation given the cluster

By definition, $X_i\mid C_i=k \sim \mathscr N(\mu_k,\Sigma_k)$, and the mean of a multivariate normal distribution is, by definition of that distribution's parameters, its own mean vector:

$$
\mathbb E[X_i\mid C_i=k]=\mu_k.
$$

$\mu_k$ is naturally interpreted as the *center* of cluster $k$ because it is precisely the average location that data points generated from that cluster would be expected to take.

Comparing the two conditional expectations: $\mathbb E[Z_i\mid X_i=x_i]$ conditions on the **observation** and asks "which cluster(s) does this point probably belong to?" — it answers a question about the point's soft membership. $\mathbb E[X_i\mid C_i=k]$ conditions on the **cluster label** and asks "where would a typical point from this cluster be located?" — it answers a question about the cluster's own geometric location. The first is a per-point diagnostic; the second is a population-level summary of the cluster.

### 6. Complete-data likelihood

If the true labels $z_i$ were observed, the complete-data likelihood is

$$
p(x_1,\dots,x_n,z_1,\dots,z_n)=\prod_{i=1}^n\prod_{k=1}^K\big[\phi_k\,\mathscr N(x_i\mid\mu_k,\Sigma_k)\big]^{z_{ik}}.
$$

Taking the log and using $\log(\prod)=\sum\log$ together with the power rule $\log(a^{z})=z\log a$:

$$
\ell_c=\log p(x_{1:n},z_{1:n})=\sum_{i=1}^n\sum_{k=1}^K z_{ik}\Big[\log\phi_k+\log\mathscr N(x_i\mid\mu_k,\Sigma_k)\Big].
$$

This expression would be easy to maximize if $z_{ik}$ were known constants, because for each fixed $i$ exactly one $z_{ik}$ equals 1 and the rest are 0 — the double sum reduces to a single sum over the $n$ points, each contributing only the log-density of its own (known) cluster, which decouples into $K$ independent, standard-form Gaussian/categorical maximum-likelihood problems (one per cluster) with closed-form solutions.

### 7. The EM interpretation

Since $z_{ik}$ is not actually observed, the EM algorithm replaces each occurrence of $z_{ik}$ in $\ell_c$ by its conditional expectation given the current parameter estimates and the observed data — precisely the responsibility $\gamma_{ik}$ derived in Part 3. This yields the expected complete-data log-likelihood

$$
Q=\sum_{i=1}^n\sum_{k=1}^K \gamma_{ik}\Big[\log\phi_k+\log\mathscr N(x_i\mid\mu_k,\Sigma_k)\Big].
$$

This is the **E-step**: rather than hard-assigning each point to a single cluster (which would require knowing the truth), it computes the full posterior/conditional distribution over cluster membership given the current model — i.e., a probabilistic ("soft") update of what is known about $C_i$ given the data and the model's current state, exactly analogous to any Bayesian conditional update.

### 8. Parameter updates (M-step)

Maximizing $Q$ with respect to $\phi_k,\mu_k,\Sigma_k$ subject to $\sum_k\phi_k=1$ yields the standard weighted maximum-likelihood updates

$$
N_k=\sum_{i=1}^n\gamma_{ik},\qquad
\phi_k^{\text{new}}=\frac{N_k}{n},\qquad
\mu_k^{\text{new}}=\frac{1}{N_k}\sum_{i=1}^n\gamma_{ik}x_i,\qquad
\Sigma_k^{\text{new}}=\frac{1}{N_k}\sum_{i=1}^n\gamma_{ik}(x_i-\mu_k^{\text{new}})(x_i-\mu_k^{\text{new}})^T.
$$

Here $\gamma_{ik}$ plays the role of a **fractional weight**: instead of each data point counting as a full, indivisible member of one cluster in the usual sample-mean/sample-covariance formulas, point $i$ contributes only a $\gamma_{ik}$-sized fraction of itself to cluster $k$'s statistics (and the remaining $1-\gamma_{ik}$ fraction is distributed across the other clusters). $N_k$ is thus the *effective* (fractional) number of points assigned to cluster $k$, and every M-step formula is simply the familiar weighted-average formula with those fractional weights plugged in.

### 9. Interpretation

Gaussian mixture clustering can be understood as a repeated cycle of conditional (Bayesian) updating rather than a one-shot geometric partitioning rule. Before seeing any particular point, the mixture weight $\phi_k$ encodes the *prior* belief that a randomly drawn observation belongs to cluster $k$. Once a specific point $x_i$ is observed, the Gaussian density $\mathscr N(x_i\mid\mu_k,\Sigma_k)$ measures how *compatible* that observation is with each cluster's current location and shape — effectively playing the role of a likelihood. Combining these two ingredients via Bayes' theorem produces the responsibility $\gamma_{ik}$, the *posterior* probability that point $i$ belongs to cluster $k$ after taking the data into account, and collecting these across all $k$ gives the soft assignment vector $\mathbb E[Z_i\mid X_i=x_i]$. The M-step then treats these posterior probabilities as fractional weights to re-estimate $\phi_k,\mu_k,\Sigma_k$, after which the whole cycle repeats with an updated prior/likelihood pair. In this light, Gaussian mixture clustering is nothing more than **probabilistic clustering built from repeated conditional expectations of a latent cluster-membership variable** — each EM iteration is one more round of "update the belief about $C_i$ given $x_i$, then update the cluster parameters given that belief.\"

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(3)
n_low, n_mid, n_high = 380, 320, 100

df = pd.DataFrame({
    "PURCHASES": np.clip(np.concatenate([
        rng.exponential(350, n_low),
        rng.normal(2300, 500, n_mid),
        rng.normal(5800, 1000, n_high),
    ]), 0, None),
    "CREDIT_LIMIT": np.clip(np.concatenate([
        rng.normal(1800, 700, n_low),
        rng.normal(6500, 1300, n_mid),
        rng.normal(11500, 1800, n_high),
    ]), 100, None),
})
df.head()
```

```python
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


class GMMFinancialSegmenter:
    # Thin wrapper around scikit-learn's GaussianMixture that adds
    # scaling, train/test splitting, out-of-sample scoring, and Plotly
    # visualization of the responsibility surface.

    def __init__(self, n_components=3, random_state=0):
        self.n_components = n_components
        self.scaler = StandardScaler()
        self.gmm = GaussianMixture(
            n_components=n_components, covariance_type="full", random_state=random_state
        )

    def split_and_scale(self, df, feature_cols, test_size=0.2, random_state=0):
        X = df[feature_cols].to_numpy()
        X_train_raw, X_test_raw = train_test_split(X, test_size=test_size, random_state=random_state)
        X_train = self.scaler.fit_transform(X_train_raw)
        X_test = self.scaler.transform(X_test_raw)
        return X_train, X_test

    def fit(self, X_train):
        self.gmm.fit(X_train)
        print(f"Converged: {self.gmm.converged_}  |  EM iterations: {self.gmm.n_iter_}")

    def test_log_likelihood(self, X_test):
        score = self.gmm.score(X_test)   # mean per-sample log-likelihood
        print(f"Average out-of-sample log-likelihood: {score:.4f}")
        return score

    def density_heatmap(self, X_train, feature_cols):
        X_orig = self.scaler.inverse_transform(X_train)
        fig = px.density_heatmap(
            x=X_orig[:, 0], y=X_orig[:, 1],
            labels={"x": feature_cols[0], "y": feature_cols[1]},
            marginal_x="histogram", marginal_y="histogram",
            title="Empirical Density of Training Data",
        )
        fig.update_traces(colorscale="Blues", selector=dict(type="histogram2d"))
        fig.update_layout(template="plotly_white")
        fig.show()

    def _responsibility_surface(self, X):
        pad = 0.5
        x_min, x_max = X[:, 0].min() - pad, X[:, 0].max() + pad
        y_min, y_max = X[:, 1].min() - pad, X[:, 1].max() + pad
        gx, gy = np.meshgrid(np.linspace(x_min, x_max, 180), np.linspace(y_min, y_max, 180))
        grid_points = np.column_stack([gx.ravel(), gy.ravel()])
        resp = self.gmm.predict_proba(grid_points)
        max_resp = resp.max(axis=1).reshape(gx.shape)
        grid_orig = self.scaler.inverse_transform(grid_points)
        gx_orig = grid_orig[:, 0].reshape(gx.shape)
        gy_orig = grid_orig[:, 1].reshape(gy.shape)
        return gx_orig, gy_orig, max_resp

    def assignment_plot(self, X, feature_cols, title):
        gx, gy, max_resp = self._responsibility_surface(X)
        labels = self.gmm.predict(X)
        X_orig = self.scaler.inverse_transform(X)

        fig = go.Figure()
        fig.add_trace(go.Contour(
            x=gx[0, :], y=gy[:, 0], z=max_resp,
            colorscale="Cividis", contours_coloring="heatmap",
            opacity=0.55, hoverinfo="skip", name="max responsibility"
        ))
        for k in range(self.n_components):
            mask = labels == k
            fig.add_trace(go.Scatter(
                x=X_orig[mask, 0], y=X_orig[mask, 1], mode="markers",
                name=f"cluster {k+1}", marker=dict(size=6, line=dict(width=0.5, color="black"))
            ))
        fig.update_layout(title=title, xaxis_title=feature_cols[0], yaxis_title=feature_cols[1],
                           template="plotly_white")
        fig.show()


features = ["PURCHASES", "CREDIT_LIMIT"]
segmenter = GMMFinancialSegmenter(n_components=3, random_state=0)
X_train, X_test = segmenter.split_and_scale(df, features, test_size=0.2, random_state=0)

segmenter.fit(X_train)
segmenter.test_log_likelihood(X_test)

segmenter.density_heatmap(X_train, features)
segmenter.assignment_plot(X_train, features, "Training Data: Soft-Assignment Confidence Surface")
segmenter.assignment_plot(X_test, features, "Test Data: Soft-Assignment Confidence Surface")
```

**Evaluation.** The density heatmap shows three overlapping clouds along the `PURCHASES`–`CREDIT_LIMIT` plane rather than one homogeneous blob, confirming the multimodal structure the synthetic data was built to have. In the training assignment plot, the contour surface forms three high-confidence "basins," and the colored markers (hard labels $\widehat C_i=\arg\max_k\gamma_{ik}$) fall almost entirely within their own basin — points landing between two basins sit in the paler, lower-confidence contour bands. The test assignment plot reuses the exact same contour surface learned only from training data, and the held-out points continue to fall mostly within their expected basins, which is visual evidence that the fitted mixture generalizes rather than having memorized the training sample (consistent with the finite, non-degenerate log-likelihood reported above).

Conceptually, the contour height at any grid location $x_{\text{grid}}$ is precisely $\max_k\gamma_k(x_{\text{grid}})=\max_k\mathbb E[Z_k\mid X=x_{\text{grid}}]$ from Part 3 — the model's soft membership expectation evaluated everywhere in feature space, not just at the observed data points. Where the surface is dark and saturated, the underlying expectation vector is close to one-hot (the model is nearly certain of membership); where it fades toward the seams between basins, the expectation vector spreads more evenly across clusters — exactly the graded, in-between membership that the soft-clustering view of Part 4 captures and that a hard assignment would simply discard.
