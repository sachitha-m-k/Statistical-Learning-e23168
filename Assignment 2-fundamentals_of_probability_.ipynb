# Q3) Transformer Failures

In a certain electricity distribution district, the number of emergency transformer failures reported in a single day is modeled as a discrete random variable $X$. Assume that $X$ takes values in the set $\{0,1,2,3\}$, with probabilities

$$P(X=0)=0.50,\qquad P(X=1)=0.30,\qquad P(X=2)=0.15,\qquad P(X=3)=0.05.$$

## Answer

**1. Probability Space:**

A suitable probability space is

$$\Omega=\{0,1,2,3\}, \qquad \mathfrak F=\mathfrak P(\Omega),$$

where $\mathfrak P(\Omega)$ is the power set of $\Omega$. Define $P$ on singletons by

$$P(\{0\})=0.50,\qquad P(\{1\})=0.30,\qquad P(\{2\})=0.15,\qquad P(\{3\})=0.05,$$

and for any event $A\subseteq \Omega$,

$$P(A)=\sum_{\omega\in A} P(\{\omega\}).$$

---

**2. Random Variable:**

Define the random variable $X:\Omega\to \mathbb R$ by

$$X(\omega)=\omega, \qquad \omega\in \Omega.$$

---

**3. Probability Mass Function:**

The probability mass function is

$$p_X(x)=P(X=x)=\begin{cases}0.50, & x=0,\\0.30, & x=1,\\0.15, & x=2,\\0.05, & x=3,\\0, & \text{otherwise}.\end{cases}$$

---

**4. Expectation:**

The expectation is

$$E[X]=\sum_x x\,p_X(x) =0(0.50)+1(0.30)+2(0.15)+3(0.05)=0.75.$$

---

**5. Variance:**

First compute

$$E[X^2]=0^2(0.50)+1^2(0.30)+2^2(0.15)+3^2(0.05)=1.35.$$

Hence

$$\mathrm{Var}(X)=E[X^2]-(E[X])^2=1.35-(0.75)^2=0.7875.$$

---

**6. Probabilities:**

The required probabilities are

$$P(X\ge 1)=P(X=1)+P(X=2)+P(X=3)=0.30+0.15+0.05=0.50,$$

and

$$P(X\ge 2)=P(X=2)+P(X=3)=0.15+0.05=0.20.$$

---

**7. Interpretation:**

The expectation $E[X]=0.75$ means that the average number of emergency transformer failures per day is $0.75$. The variance $\mathrm{Var}(X)=0.7875$ measures the day-to-day variability in the number of failures, so it indicates how much actual daily counts may fluctuate around the average.

> In practice, this discrete model helps the utility plan resources and staffing for emergency response, based on typical daily failure rates.

```python
import numpy as np
import matplotlib.pyplot as plt

x_vals = [0, 1, 2, 3]
probs = [0.50, 0.30, 0.15, 0.05]

# Verify probabilities sum to 1
print(f"Sum of probabilities: {sum(probs)}")

# Calculate E[X] and Var(X)
E_X = sum(x * p for x, p in zip(x_vals, probs))
E_X2 = sum(x**2 * p for x, p in zip(x_vals, probs))
Var_X = E_X2 - E_X**2

print(f"E[X] = {E_X}")
print(f"E[X²] = {E_X2}")
print(f"Var(X) = {Var_X}")
print(f"σ = {np.sqrt(Var_X):.4f}")

# Probabilities
print(f"\nP(X ≥ 1) = {sum(probs[1:])}")
print(f"P(X ≥ 2) = {sum(probs[2:])}")

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# PMF
axes[0].bar(x_vals, probs, color='steelblue', edgecolor='black', width=0.6)
axes[0].set_xlabel('Number of failures (x)', fontsize=11)
axes[0].set_ylabel('P(X = x)', fontsize=11)
axes[0].set_title('Probability Mass Function of X', fontsize=12)
axes[0].set_xticks(x_vals)
for i, p in enumerate(probs):
    axes[0].text(i, p + 0.01, f'{p:.2f}', ha='center', fontsize=10)

# CDF
cdf = np.cumsum(probs)
axes[1].step(x_vals, cdf, where='post', color='darkred', lw=2, label='CDF')
axes[1].scatter(x_vals, cdf, color='darkred', s=50, zorder=5)
axes[1].set_xlabel('Number of failures (x)', fontsize=11)
axes[1].set_ylabel('P(X ≤ x)', fontsize=11)
axes[1].set_title('Cumulative Distribution Function', fontsize=12)
axes[1].set_xticks(x_vals)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

# Q4) Probability of Defective Items

In a manufacturing process, the proportion of defective items in a large batch produced on a given day is modeled by a continuous random variable $X$. Since $X$ represents a proportion, it takes values in the interval $[0,1]$. Suppose that $X$ has a Beta distribution with parameters $\alpha=2$ and $\beta=18$, so that its density is

$$f(x)=\begin{cases}\dfrac{1}{B(2,18)}x^{1}(1-x)^{17}, & 0<x<1,\\0, & \text{otherwise},\end{cases}$$

where $B(2,18)$ is the Beta function.

## Answer

**1. Probability Space:**

A suitable probability space is

$$\Omega=[0,1], \qquad \mathfrak F=\mathfrak B([0,1]),$$

where $\mathfrak B([0,1])$ is the Borel $\sigma$-algebra on $[0,1]$. Define the probability measure $P$ by

$$P(A)=\int_A f(x)\,dx, \qquad A\in \mathfrak F,$$

where

$$f(x)=\begin{cases}\dfrac{1}{B(2,18)}x(1-x)^{17}, & 0<x<1,\\0, & \text{otherwise}.\end{cases}$$

---

**2. Random Variable:**

Define the random variable $X:\Omega\to\mathbb R$ by

$$X(\omega)=\omega, \qquad \omega\in[0,1].$$

Thus $X$ records the defective proportion in the batch.

---

**3. Probability Density Function:**

The probability density function of $X$ is

$$f_X(x)=\begin{cases}\dfrac{1}{B(2,18)}x(1-x)^{17}, & 0<x<1,\\0, & \text{otherwise}.\end{cases}$$

---

**4. Valid Density:**

To verify that $f$ is a valid density, note first that $f(x)\ge 0$ for all $x$. Also,

$$\int_{-\infty}^{\infty} f(x)\,dx =\int_0^1 \dfrac{1}{B(2,18)}x(1-x)^{17}\,dx =\dfrac{1}{B(2,18)}B(2,18) =1.$$

So $f$ is a valid probability density function.

---

**5. Expectation:**

Since $X\sim \mathrm{Beta}(2,18)$,

$$E[X]=\frac{\alpha}{\alpha+\beta}=\frac{2}{20}=0.10.$$

---

**6. Variance:**

The variance of a Beta$(\alpha,\beta)$ random variable is

$$\mathrm{Var}(X)=\frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)}.$$

Hence

$$\mathrm{Var}(X)=\frac{2\cdot 18}{20^2\cdot 21} =\frac{36}{8400} =\frac{3}{700} \approx 0.00429.$$

---

**7. Probability $P(X \le 0.10)$:**

The probability that the defective proportion is at most $0.10$ is

$$P(X\le 0.10)=\int_0^{0.10} \dfrac{1}{B(2,18)}x(1-x)^{17}\,dx.$$

Equivalently,

$$P(X\le 0.10)=I_{0.10}(2,18),$$

where $I_x(\alpha,\beta)$ denotes the regularized incomplete Beta function.

---

**8. Probability $P(0.05 \le X \le 0.15)$:**

The probability that the defective proportion lies between $0.05$ and $0.15$ is

$$P(0.05\le X\le 0.15)=\int_{0.05}^{0.15} \dfrac{1}{B(2,18)}x(1-x)^{17}\,dx.$$

Equivalently,

$$P(0.05\le X\le 0.15)=I_{0.15}(2,18)-I_{0.05}(2,18).$$

---

**9. Why Beta?**

A Beta distribution is reasonable here because the defective proportion is a continuous quantity constrained to lie between $0$ and $1$, and the Beta family is specifically designed to model proportions and rates.

---

**10. Interpretation:**

The expectation

$$E[X]=0.10$$

means that the long-run average defective proportion is $10\%$. The variance

$$\mathrm{Var}(X)\approx 0.00429$$

measures the variability in the defective proportion from batch to batch, so it indicates how stable or unstable the manufacturing quality is over time.

> A small variance indicates reliable, consistent quality; a large variance indicates that batches are unpredictable, suggesting a need for process improvement.

```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt

alpha, beta_param = 2, 18

# Mean and variance
mean = alpha / (alpha + beta_param)
variance = (alpha * beta_param) / ((alpha + beta_param)**2 * (alpha + beta_param + 1))
std_dev = np.sqrt(variance)

print(f"Beta({alpha}, {beta_param})")
print(f"E[X] = {mean:.4f}")
print(f"Var(X) = {variance:.6f}")
print(f"σ = {std_dev:.6f}")

# Probabilities
p_le_010 = beta.cdf(0.10, alpha, beta_param)
p_le_005 = beta.cdf(0.05, alpha, beta_param)
p_le_015 = beta.cdf(0.15, alpha, beta_param)

print(f"\nP(X ≤ 0.10) = {p_le_010:.4f}")
print(f"P(0.05 ≤ X ≤ 0.15) = {p_le_015 - p_le_005:.4f}")

# Visualization
x = np.linspace(0, 1, 1000)
y = beta.pdf(x, alpha, beta_param)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y, 'b-', lw=2.5, label=f'Beta({alpha}, {beta_param})')
ax.fill_between(x[x <= 0.10], y[x <= 0.10], alpha=0.3, color='green', label=f'P(X ≤ 0.10) = {p_le_010:.3f}')
ax.fill_between(x[(x >= 0.05) & (x <= 0.15)], y[(x >= 0.05) & (x <= 0.15)], alpha=0.3, color='orange', label=f'P(0.05 ≤ X ≤ 0.15) = {p_le_015 - p_le_005:.3f}')
ax.axvline(mean, color='red', linestyle='--', lw=2, label=f'E[X] = {mean:.2f}')
ax.set_xlabel('Defective proportion (x)', fontsize=11)
ax.set_ylabel('Probability density', fontsize=11)
ax.set_title(f'Beta Distribution of Defective Proportion', fontsize=12)
ax.legend(fontsize=10)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

# Q5) Expected Project Completion Time

A construction company is estimating the time required to install a backup generator at a hospital. Let $X$ denote the number of hours needed to complete the installation. Based on expert judgment, the shortest possible time is $4$ hours, the longest possible time is $10$ hours, and the most likely completion time is $6$ hours. Assume that $X$ follows a Triangular distribution with parameters $a=4$, $b=10$, and $c=6$.

## Answer

**1. Support:**

Since $X$ represents the installation time in hours, and the best-case and worst-case estimates are $4$ and $10$, the support of $X$ is

$$4 \leq X \leq 10.$$

---

**2. Probability Density Function:**

Since $X \sim \mathrm{Triangular}(4,10,6)$, its density is

$$f_X(x)=\begin{cases}\dfrac{x-4}{6}\, & 4 \leq x < 6,\\\dfrac{10-x}{12}\, & 6 < x \leq 10,\\0, & \text{otherwise}.\end{cases}$$

---

**3. Valid Density:**

To verify that $f_X$ is a density, compute

$$\int_{4}^{6} \frac{x-4}{6}\,dx + \int_{6}^{10} \frac{10-x}{12}\,dx.$$

For the first part,

$$\int_{4}^{6} \frac{x-4}{6}\,dx = \frac{1}{6}\cdot \frac{(6-4)^2}{2} = \frac{1}{3}.$$

For the second part,

$$\int_{6}^{10} \frac{10-x}{12}\,dx = \frac{1}{12}\cdot \frac{(10-6)^2}{2} = \frac{2}{3}.$$

Hence

$$\int_{-\infty}^{\infty} f_X(x)\,dx = \frac{1}{3}+\frac{2}{3}=1.$$ ✓

---

**4. Expectation:**

The expectation of a Triangular distribution is

$$E[X]=\frac{a+b+c}{3}=\frac{4+10+6}{3}=\frac{20}{3}\approx 6.67.$$

---

**5. Variance:**

The variance is

$$\mathrm{Var}(X)=\frac{a^2+b^2+c^2-ab-ac-bc}{18}=\frac{16+100+36-40-24-60}{18}=\frac{28}{18}=\frac{14}{9}\approx 1.56.$$

---

**6. Parameters:**

In this context:

* $a=4$ is the shortest possible installation time,
* $b=10$ is the longest possible installation time,
* $c=6$ is the most likely installation time.

---

**7. Why Triangular?**

A Triangular distribution is reasonable here because the company may not have enough past data for a more elaborate statistical model, but experts can still provide a minimum, maximum, and most likely estimate. The distribution is bounded and reflects these three practical judgments.

---

**8. Probability $P(X \leq 5)$:**

To compute $P(X\leq 5)$, use the first branch of the density:

$$P(X\leq 5)=\int_{4}^{5} \frac{x-4}{6}\,dx = \frac{1}{6}\cdot\frac{(5-4)^2}{2} = \frac{1}{12}\approx 0.0833.$$

---

**9. Probability $P(5 \leq X \leq 8)$:**

To compute $P(5\leq X\leq 8)$, split the integral at $x=6$:

$$P(5\leq X\leq 8)=\int_{5}^{6} \frac{x-4}{6}\,dx + \int_{6}^{8} \frac{10-x}{12}\,dx = \frac{1}{4}+\frac{1}{2}=\frac{3}{4}.$$

---

**10. Applications:**

This model can be used in project planning by estimating the likelihood that the installation will finish within a given time window. It can also be used in sensitivity analysis by showing how uncertainty in task duration affects schedules, staffing, and contingency planning.

> The Triangular distribution is practical for expert elicitation when only three-point estimates are available, making it popular in PERT and project risk analysis.

```python
import numpy as np
from scipy.stats import triang
import matplotlib.pyplot as plt

a, b, c = 4, 10, 6
c_scipy = (c - a) / (b - a)  # Scipy parameterization

# Mean and variance
mean = (a + b + c) / 3
variance = (a**2 + b**2 + c**2 - a*b - a*c - b*c) / 18
std_dev = np.sqrt(variance)

print(f"Triangular({a}, {b}, {c})")
print(f"E[X] = {mean:.4f}")
print(f"Var(X) = {variance:.4f}")
print(f"σ = {std_dev:.4f}")

# Probabilities
p_le_5 = triang.cdf(5, c=c_scipy, loc=a, scale=b-a)
p_5_8 = triang.cdf(8, c=c_scipy, loc=a, scale=b-a) - triang.cdf(5, c=c_scipy, loc=a, scale=b-a)

print(f"\nP(X ≤ 5) = {p_le_5:.4f}")
print(f"P(5 ≤ X ≤ 8) = {p_5_8:.4f}")

# Visualization
x = np.linspace(a-1, b+1, 1000)
y = triang.pdf(x, c=c_scipy, loc=a, scale=b-a)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y, 'b-', lw=2.5, label=f'Triangular({a}, {b}, {c})')
ax.fill_between(x[x <= 5], y[x <= 5], alpha=0.3, color='green', label=f'P(X ≤ 5) = {p_le_5:.4f}')
ax.fill_between(x[(x >= 5) & (x <= 8)], y[(x >= 5) & (x <= 8)], alpha=0.3, color='orange', label=f'P(5 ≤ X ≤ 8) = {p_5_8:.4f}')
ax.axvline(mean, color='red', linestyle='--', lw=2, label=f'E[X] = {mean:.2f}')
ax.axvline(c, color='purple', linestyle=':', lw=2, label=f'Mode = {c}')
ax.set_xlabel('Installation time (hours)', fontsize=11)
ax.set_ylabel('Probability density', fontsize=11)
ax.set_title('Triangular Distribution of Installation Time', fontsize=12)
ax.legend(fontsize=10)
ax.grid(alpha=0.3)
ax.set_xlim(a-1, b+1)
plt.tight_layout()
plt.show()
```

# Q6) Expected Project Completion Time Using PERT

A project manager is estimating the time required to complete a critical software integration task. Let $X$ denote the number of days required to complete the task. Based on expert judgment, the shortest possible time is $a=8$ days, the longest possible time is $b=20$ days, and the most likely completion time is $c=11$ days. Assume that $X$ follows a PERT distribution with parameters $(a,b,c)$.

## Answer

**1. Shape Parameters:**

The PERT shape parameters are

$$\alpha = 1 + 4\left( \frac{c-a}{b-a} \right) = 1 + 4\left( \frac{11-8}{20-8} \right) = 1 + 4\left( \frac{3}{12} \right) = 2$$

and

$$\beta = 1 + 4\left( \frac{b-c}{b-a} \right) = 1 + 4\left( \frac{20-11}{20-8} \right) = 1 + 4\left( \frac{9}{12} \right) = 4.$$

---

**2. Support:**

Since $X$ represents the task duration in days, its support is

$$8 \leq X \leq 20.$$

---

**3. Probability Density Function:**

The probability density function is

$$f_X(x)=\frac{(x-8)(20-x)^3}{B(2,4)\,12^5}, \qquad 8 \leq x \leq 20.$$

Since $B(2,4)=\frac{1!\,3!}{5!}=\frac{1}{20}$,

$$f_X(x)=\frac{20(x-8)(20-x)^3}{12^5}, \qquad 8 \leq x \leq 20.$$

---

**4. Probability Space:**

A suitable probability space is

$$\Omega=[8,20], \qquad \mathfrak F=\mathfrak B([8,20]),$$

where $\mathfrak B([8,20])$ is the Borel $\sigma$-algebra on $[8,20]$. Define the probability measure $P$ by

$$P(A)=\int_A f_X(x)\,dx, \qquad A\in \mathfrak F.$$

---

**5. Random Variable:**

Define the random variable $X:\Omega\to\mathbb R$ by

$$X(\omega)=\omega, \qquad \omega\in[8,20].$$

Thus $X$ records the task duration in days.

---

**6. Expectation:**

The PERT expectation is

$$E[X]=\frac{a+4c+b}{6}=\frac{8+4(11)+20}{6}=\frac{72}{6}=12.$$

---

**7. Variance:**

The variance approximation is

$$\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7}=\frac{(12-8)(20-12)}{7}=\frac{32}{7}\approx 4.57.$$

---

**8. Standard Deviation:**

The standard deviation is

$$\sigma=\sqrt{\mathrm{Var}(X)}=\sqrt{\frac{32}{7}}\approx 2.14.$$

---

**9. Comparison with Triangular:**

| Quantity | Triangular | PERT |
|----------|-----------|------|
| Mean | $\approx 6.67$ | $12$ |
| Variance | $\approx 1.56$ | $\approx 4.57$ |
| Std. Deviation | $\approx 1.25$ | $\approx 2.14$ |

The PERT model is more appropriate when we want the expected value to be closer to the mode (most likely case), reflecting expert confidence in that estimate. PERT uses a smooth Beta distribution weighted toward the mode, whereas Triangular uses linear ramps and is less peaked.

---

**10. Interpretation:**

In this setting:

* $a=8$ is the shortest plausible completion time,
* $b=20$ is the longest plausible completion time,
* $c=11$ is the most likely completion time,
* $E[X]=12$ is the expected task duration,
* $\sigma \approx 2.14$ measures the uncertainty or variability in the task duration.

In project planning, these values help managers estimate realistic schedules, assess deadline risk, and decide whether contingency time should be included.

> PERT is widely used in project management because it concentrates probability near the most likely estimate, reflecting expert judgment more faithfully than a uniform or purely symmetric distribution.

```python
import numpy as np
from scipy.stats import triang, beta
import matplotlib.pyplot as plt

a, b, c = 8, 20, 11

# PERT parameters
alpha_pert = 1 + 4 * (c - a) / (b - a)
beta_pert = 1 + 4 * (b - c) / (b - a)

# PERT statistics
mean_pert = (a + 4*c + b) / 6
var_pert = (mean_pert - a) * (b - mean_pert) / 7
std_pert = np.sqrt(var_pert)

print(f"PERT Distribution (α={alpha_pert}, β={beta_pert})")
print(f"E[X] = {mean_pert:.4f}")
print(f"Var(X) = {var_pert:.4f}")
print(f"σ = {std_pert:.4f}")

# Compare with Triangular
a_tri, b_tri, c_tri = 4, 10, 6
mean_tri = (a_tri + b_tri + c_tri) / 3
var_tri = (a_tri**2 + b_tri**2 + c_tri**2 - a_tri*b_tri - a_tri*c_tri - b_tri*c_tri) / 18
std_tri = np.sqrt(var_tri)

print(f"\nComparison:")
print(f"Triangular:  E[X]={mean_tri:.4f}, σ={std_tri:.4f}")
print(f"PERT:        E[X]={mean_pert:.4f}, σ={std_pert:.4f}")

# Visualization
x = np.linspace(a, b, 500)

c_scipy_tri = (c_tri - a_tri) / (b_tri - a_tri)
y_tri = triang.pdf(x, c=c_scipy_tri, loc=a_tri, scale=b_tri-a_tri)
y_pert = beta.pdf(x, alpha_pert, beta_pert, loc=a, scale=b-a)

fig, ax = plt.subplots(figsize=(11, 5))
ax.plot(x, y_tri, 'b--', lw=2.5, label=f'Triangular (E={mean_tri:.2f})', alpha=0.7)
ax.plot(x, y_pert, 'r-', lw=2.5, label=f'PERT (E={mean_pert:.2f})')
ax.axvline(c, color='purple', linestyle=':', lw=2, label=f'Most likely = {c}')
ax.set_xlabel('Task duration (days)', fontsize=11)
ax.set_ylabel('Probability density', fontsize=11)
ax.set_title('Comparison: Triangular vs PERT Distribution', fontsize=12)
ax.legend(fontsize=10)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

# Q7) If the PERT Expectation is to Match Our Intuition

A project manager models the duration $X$ of a task by a PERT distribution with parameters $a$, $b$, and $c$, where $a$ is the minimum time, $b$ is the maximum time, and $c$ is the most likely time.

Suppose the minimum time is $a=5$ days and the most likely time is $c=11$ days. If the expected time computed from the PERT model is also to be the same as the most likely time, what is the value of the maximum time $b$?

## Answer

**1. Determining $b$:**

Use the PERT mean formula

$$E[X]=\frac{a+4c+b}{6}.$$

If $E[X]=c$, then

$$\frac{a+4c+b}{6}=c$$

which implies

$$a+4c+b=6c$$

and therefore

$$b=2c-a=2(11)-5=\boxed{17}.$$

---

**2. What Happens to the Variance?**

Using the approximation

$$\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7},$$

and imposing $E[X]=c$, we get

$$\mathrm{Var}(X)=\frac{(c-a)(b-c)}{7}.$$

Since $b=2c-a$, this becomes

$$\mathrm{Var}(X)=\frac{(c-a)^2}{7}=\frac{(11-5)^2}{7}=\frac{36}{7}\approx 5.14.$$

If the expected value equals the most likely value, then the distribution becomes **symmetric** about the mode:
- $c - a = 11 - 5 = 6$
- $b - c = 17 - 11 = 6$

The variance simplifies to $(c-a)^2/7$, showing that variance depends only on the "half-width" of the estimates around the mode.

> When $E[X]=c$, the PERT distribution becomes perfectly symmetric about the mode. The variance depends solely on the "half-width" of the range: $(c-a)^2/7$. This symmetric case is pedagogically important because it shows the special relationship between the three estimates when the mean equals the mode.

```python
import numpy as np

a, c = 5, 11
b = 2*c - a

print(f"Given: a = {a}, c = {c}")
print(f"Constraint: E[X] = c")
print(f"\nFrom E[X] = (a + 4c + b) / 6 = c:")
print(f"  a + 4c + b = 6c")
print(f"  b = 2c - a = 2({c}) - {a} = {b}")

# Mean and variance
mean = (a + 4*c + b) / 6
variance = (mean - a) * (b - mean) / 7
std_dev = np.sqrt(variance)

print(f"\nVerification:")
print(f"  E[X] = ({a} + 4({c}) + {b}) / 6 = {mean:.4f}")
print(f"  E[X] = c? {np.isclose(mean, c)}")

# Check symmetry
print(f"\nSymmetry check:")
print(f"  c - a = {c - a}")
print(f"  b - c = {b - c}")
print(f"  Symmetric? {c - a == b - c}")

print(f"\nVariance:")
print(f"  Var(X) = (c - a)² / 7 = {(c - a)**2} / 7 = {variance:.4f}")
print(f"  σ = {std_dev:.4f}")
```
