# Maximum Entropy Inference — Assignment Solutions


This notebook works through all four assignment questions using Jaynes' **Principle of Maximum Entropy (MaxEnt)** as the unifying framework: in every problem we maximize an entropy functional subject to whatever macroscopic/empirical constraints we are given, form the Lagrangian, and read off the generalized Boltzmann/Gibbs form
$$p^*(x) \propto \exp\left(-\sum_i \lambda_i f_i(x)\right),$$
then either solve analytically for the Lagrange multipliers (Q1, Q2) or numerically (Q3, Q4).


# Q1. Statistical Mechanics of the Quantum Oscillator

## 1.1 Setup


A single quantum harmonic oscillator in thermal equilibrium with a bath at temperature $T$ has energy eigenvalues
$$E_j = \hbar\omega\left(j+\tfrac12\right), \qquad j = 0,1,2,\dots$$

This is a **closed system**: only energy (not particle number) is exchanged with the bath, so the single macroscopic constraint is the mean energy $U = \mathbb{E}[E]$. From the general MaxEnt result for the Canonical Ensemble (Section 2 of the lecture notes), the maximum-entropy microstate distribution is
$$p_j^* = \frac{1}{Z}e^{-\beta E_j}, \qquad \beta \triangleq \frac{1}{k_BT}, \qquad Z(T) = \sum_{j=0}^\infty e^{-\beta E_j}.$$


## 1.2 The Partition Function


Substituting the eigenvalues,
$$Z(T) = \sum_{j=0}^\infty e^{-\beta\hbar\omega(j+\frac12)} = e^{-\beta\hbar\omega/2}\sum_{j=0}^\infty \left(e^{-\beta\hbar\omega}\right)^j .$$

The sum is a geometric series in $r \triangleq e^{-\beta\hbar\omega}$. Since $\beta,\hbar,\omega>0$ we have $0<r<1$, so the series **converges** to $\dfrac{1}{1-r}$. Hence

$$\boxed{Z(T) = \frac{e^{-\hbar\omega/(2k_BT)}}{1-e^{-\hbar\omega/(k_BT)}}}$$


## 1.3 Thermodynamic Potentials


**Helmholtz Free Energy.** By definition $F = -k_BT\ln Z$, so
$$F(T) = -k_BT\left[-\frac{\hbar\omega}{2k_BT} - \ln\!\left(1-e^{-\hbar\omega/(k_BT)}\right)\right]
= \frac{\hbar\omega}{2} + k_BT\ln\!\left(1-e^{-\hbar\omega/(k_BT)}\right).$$

**Internal Energy.** Using $U = -\dfrac{\partial \ln Z}{\partial \beta}$, write $\ln Z = -\tfrac{\beta\hbar\omega}{2} - \ln\!\left(1-e^{-\beta\hbar\omega}\right)$. Then
$$\frac{\partial \ln Z}{\partial \beta} = -\frac{\hbar\omega}{2} - \frac{\hbar\omega\, e^{-\beta\hbar\omega}}{1-e^{-\beta\hbar\omega}} = -\frac{\hbar\omega}{2}-\frac{\hbar\omega}{e^{\beta\hbar\omega}-1}.$$

So
$$\boxed{U(T) = \frac{\hbar\omega}{2} + \frac{\hbar\omega}{e^{\hbar\omega/(k_BT)}-1}}$$

which is exactly Planck's mean-energy formula: the zero-point energy $\hbar\omega/2$ plus $\hbar\omega\,\bar n(T)$, where $\bar n(T)=(e^{\beta\hbar\omega}-1)^{-1}$ is the Bose-Einstein occupation number.


## 1.4 Entropy


From $S = -\left(\dfrac{\partial F}{\partial T}\right)$, differentiate
$$F(T) = \frac{\hbar\omega}{2} + k_BT\ln\!\left(1-e^{-\hbar\omega/(k_BT)}\right).$$

Let $x \triangleq \hbar\omega/(k_BT)$. Then $F = \tfrac{\hbar\omega}{2} + k_BT\ln(1-e^{-x})$, and using the product rule together with $\dfrac{dx}{dT}=-\dfrac{x}{T}$:
$$\frac{\partial F}{\partial T} = k_B\ln(1-e^{-x}) + k_BT\cdot\frac{e^{-x}}{1-e^{-x}}\cdot\left(-\frac{dx}{dT}\right) = k_B\ln(1-e^{-x}) - k_B\,\frac{x\,e^{-x}}{1-e^{-x}}.$$

Hence
$$\boxed{S(T) = -k_B\ln\!\left(1-e^{-\hbar\omega/(k_BT)}\right) + k_B\,\frac{\hbar\omega/(k_BT)}{e^{\hbar\omega/(k_BT)}-1}}$$

**Check that $U = F+TS$:**
$$F+TS = \frac{\hbar\omega}{2}+k_BT\ln(1-e^{-x}) - k_BT\ln(1-e^{-x}) + k_BT\cdot\frac{x}{e^x-1} = \frac{\hbar\omega}{2}+\frac{\hbar\omega}{e^{\hbar\omega/(k_BT)}-1} = U(T).\ \checkmark$$


## 1.5 Heat Capacity and Hyperbolic Form


Differentiating $U(T)=\dfrac{\hbar\omega}{2}+\dfrac{\hbar\omega}{e^x-1}$ (with $x=\hbar\omega/(k_BT)$) with respect to $T$:

$$C_V = \frac{\partial U}{\partial T} = \hbar\omega\cdot\left(-\frac{e^{x}}{(e^{x}-1)^2}\right)\frac{dx}{dT} = \hbar\omega\cdot\frac{e^{x}}{(e^{x}-1)^2}\cdot\frac{x}{T} = k_B x^2\,\frac{e^{x}}{(e^{x}-1)^2}.$$

To reach the hyperbolic form, multiply numerator and denominator by $e^{-x}$:
$$\frac{e^{x}}{(e^{x}-1)^2} = \frac{1}{(e^{x/2}-e^{-x/2})^2} = \frac{1}{4\sinh^2(x/2)},$$
using $\sinh(u) = \tfrac12(e^{u}-e^{-u})$ with $u=x/2$. Therefore

$$\boxed{C_V = k_B\left(\frac{\hbar\omega}{2k_BT}\right)^2\frac{1}{\sinh^2\!\left(\dfrac{\hbar\omega}{2k_BT}\right)}}$$

which is the **Einstein heat-capacity function**.


## 1.6 High-Temperature Limit


As $k_BT \gg \hbar\omega$, $x = \hbar\omega/(k_BT) \to 0$. Using $\sinh(x/2)\approx x/2$ for small $x$,
$$C_V \approx k_B\, x^2 \cdot \frac{1}{(x/2)^2} = k_B\, x^2\cdot\frac{4}{x^2} = 4k_B \cdot \frac14 \cdot ... $$

More carefully: $\sinh^2(x/2) \approx (x/2)^2\left[1+\tfrac{(x/2)^2}{3}+\dots\right]^2 \approx (x/2)^2$ to leading order, so
$$C_V \approx k_B x^2 \cdot \frac{1}{(x/2)^2} = k_B x^2 \cdot \frac{4}{x^2} = 4k_B\cdot\frac{x^2}{x^2}\Big/1 = k_B.$$

So $C_V \to k_B$ in the high-temperature limit — **this matches the classical equipartition theorem**, which assigns $\tfrac12 k_B$ to each quadratic degree of freedom in the Hamiltonian $H=\tfrac{p^2}{2m}+\tfrac12 m\omega^2 q^2$ (kinetic + potential), for a total of $2\times\tfrac12 k_B = k_B$. Quantum effects (the zero-point energy and the freezing-out of $C_V\to0$ as $T\to0$) only become visible when $k_BT$ is comparable to or smaller than the level spacing $\hbar\omega$.


```python
import numpy as np
import plotly.graph_objects as go

hbar_omega = 1.0   # set hbar*omega = 1 (natural units); kB = 1
kB = 1.0

x = np.linspace(0.02, 6, 800)                 # x = hbar*omega/(kB*T)
Cv = kB * (x/2)**2 / np.sinh(x/2)**2

fig = go.Figure()
fig.add_trace(go.Scatter(x=1/x, y=Cv, mode='lines', name=r'$C_V(T)$',
                          line=dict(color='#1f77b4', width=3)))
fig.add_hline(y=1.0, line=dict(color='#d62728', dash='dash'),
              annotation_text='Classical equipartition limit  C_V = k_B', annotation_position='bottom right')
fig.update_layout(title='<b>Einstein Heat Capacity of a Quantum Oscillator</b>',
                   xaxis_title='k_B T / (hbar omega)',
                   yaxis_title='C_V / k_B',
                   template='plotly_white', width=820, height=480)
fig.show()
```

# Q2. The Maximum Entropy Distribution under Mean and Variance Constraints

## 2.1 Setup


Let $X\in(-\infty,\infty)$ with reference measure $m(x)=1$ (flat prior on the whole real line). We are given two constraints:
$$\mathbb{E}[X] = \mu, \qquad \mathbb{E}[(X-\mu)^2] = \sigma^2.$$

We seek $p(x)$ maximizing the continuous Shannon-Jaynes entropy
$$H_c[p] = -\int_{-\infty}^{\infty} p(x)\ln p(x)\,dx$$
subject to normalization and the two moment constraints above.


## 2.2 The Lagrangian and MaxEnt Form


Form the Lagrangian with multipliers $\gamma$ (normalization), $\lambda_1$ (mean) and $\lambda_2$ (variance):
$$\mathscr{L}[p] = -\int p\ln p\,dx - \gamma\left(\int p\,dx - 1\right) - \lambda_1\left(\int p\,x\,dx-\mu\right) - \lambda_2\left(\int p\,(x-\mu)^2\,dx - \sigma^2\right).$$

Taking the functional derivative $\delta \mathscr{L}/\delta p(x)=0$ pointwise gives
$$-(1+\ln p(x)) - \gamma - \lambda_1 x - \lambda_2(x-\mu)^2 = 0
\quad\Longrightarrow\quad
p^*(x) = \frac{1}{Z}\exp\left(-\lambda_1 x - \lambda_2(x-\mu)^2\right),$$
where $Z=e^{1+\gamma}$ is the partition function (normalization constant).


## 2.3 Solving for the Multipliers


Complete the square in the exponent. Write $\lambda_1 x = \lambda_1(x-\mu) + \lambda_1\mu$, so
$$-\lambda_1 x - \lambda_2(x-\mu)^2 = -\lambda_2\left[(x-\mu)^2 + \frac{\lambda_1}{\lambda_2}(x-\mu)\right] - \lambda_1\mu
= -\lambda_2\left(x-\mu+\frac{\lambda_1}{2\lambda_2}\right)^2 + \frac{\lambda_1^2}{4\lambda_2} - \lambda_1\mu.$$

For $p^*(x)$ to be centered at $\mu$ (as required by $\mathbb{E}[X]=\mu$), the linear shift inside the square must vanish, forcing
$$\lambda_1 = 0.$$

With $\lambda_1=0$, the density reduces to $p^*(x) = \frac1Z\exp(-\lambda_2(x-\mu)^2)$, a Gaussian kernel. The standard Gaussian integral
$$\int_{-\infty}^\infty e^{-\lambda_2(x-\mu)^2}\,dx = \sqrt{\frac{\pi}{\lambda_2}} \implies Z = \sqrt{\pi/\lambda_2},$$
and matching the second moment,
$$\int_{-\infty}^\infty (x-\mu)^2\, p^*(x)\,dx = \frac{1}{2\lambda_2} \stackrel{!}{=} \sigma^2 \implies \lambda_2 = \frac{1}{2\sigma^2}.$$

So
$$\lambda_1 = 0, \qquad \lambda_2 = \frac{1}{2\sigma^2}, \qquad Z = \sqrt{2\pi\sigma^2}.$$


## 2.4 Result


Substituting back,
$$\boxed{p^*(x) = \frac{1}{\sqrt{2\pi\sigma^2}}\exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)}$$

i.e. $X\sim\mathcal{N}(\mu,\sigma^2)$. This is the precise information-theoretic sense in which the **Gaussian is the least-biased, "maximally honest" distribution** consistent with knowing only a mean and a variance: any other distribution with the same first two moments smuggles in extra, unjustified structure (skew, multimodality, heavier/lighter tails) that the data given does not support.


```python
import sympy as sp

x, mu, sigma, lam1, lam2, gamma = sp.symbols('x mu sigma lambda_1 lambda_2 gamma', positive=False, real=True)
sigma = sp.symbols('sigma', positive=True)

p_star = sp.exp(-lam1*x - lam2*(x-mu)**2)

# Solve lambda1 = 0, lambda2 = 1/(2 sigma^2) by matching moments symbolically
lam2_val = 1/(2*sigma**2)
p_sub = p_star.subs({lam1:0, lam2:lam2_val})

Z = sp.integrate(p_sub, (x, -sp.oo, sp.oo))
mean_check  = sp.integrate(x*p_sub, (x, -sp.oo, sp.oo)) / Z
var_check   = sp.integrate((x-mu)**2*p_sub, (x, -sp.oo, sp.oo)) / Z

print("Z            =", sp.simplify(Z))
print("E[X]         =", sp.simplify(mean_check))
print("Var[X]       =", sp.simplify(var_check))
```

# Q3. Multiclass (Binary) Logistic Regression as a Conditional MaxEnt Classifier

## 3.1 Data


The AI Feature Extractor Agent emits binary features $\mathbf{x}=(x_1,x_2,x_3)^T$ (financial / account-access / bug-technical context flags). Five scenarios occur with fixed marginal probabilities:

| Scenario | $x_1$ | $x_2$ | $x_3$ | $P(\mathbf{x})$ |
|---|---|---|---|---|
| A — Pure Financial | 1 | 0 | 0 | 0.40 |
| B — Pure Access | 0 | 1 | 0 | 0.25 |
| C — Pure Technical | 0 | 0 | 1 | 0.05 |
| D — Access + Technical | 0 | 1 | 1 | 0.15 |
| E — Billing + Access | 1 | 1 | 0 | 0.05 |

and we are given three empirical joint-expectation constraints on the route to the **Technical** desk, $Y=1$:
$$\mathbb{E}[x_1\cdot\mathbb{I}(Y=1)] = 0.05, \quad \mathbb{E}[x_2\cdot\mathbb{I}(Y=1)] = 0.15, \quad \mathbb{E}[x_3\cdot\mathbb{I}(Y=1)] = 0.16.$$


## 3.2 Conditional MaxEnt Derivation


We do not know $P(Y=1\mid\mathbf{x})$ for each scenario, only these three *aggregate* moments. Following Section 1 of the notes, we pick the **least-biased** conditional law by maximizing the conditional entropy
$$H(Y\mid \mathbf{x}) = -\sum_{\mathbf{x}}P(\mathbf{x})\sum_{y\in\{0,1\}}p(y\mid\mathbf{x})\ln p(y\mid \mathbf{x})$$
subject to, for each scenario, $\sum_y p(y\mid\mathbf{x})=1$, and globally,
$$\sum_{\mathbf{x}}P(\mathbf{x})\, x_i\, p(Y=1\mid \mathbf{x}) = \langle x_i\,\mathbb{I}(Y=1)\rangle, \qquad i=1,2,3.$$

Introduce Lagrange multipliers $w_1,w_2,w_3$ for the three global constraints and $\gamma(\mathbf{x})$ for each per-$\mathbf x$ normalization. Differentiating the Lagrangian with respect to $p(y\mid\mathbf x)$ and setting it to zero gives, for $y=1$,
$$p(1\mid\mathbf{x}) \propto \exp\!\left(\sum_i w_i x_i\right), \qquad\text{and for } y=0,\quad p(0\mid \mathbf x)\propto \exp(0)=1,$$
since only the $Y=1$ indicator appears in the constraint features. Normalizing over $y\in\{0,1\}$:

$$\boxed{p(Y=1\mid \mathbf{x}) = \frac{e^{\mathbf{w}\cdot\mathbf{x}}}{1+e^{\mathbf{w}\cdot\mathbf{x}}} = \operatorname{sigmoid}(\mathbf w\cdot \mathbf x)}$$

This is exactly **binary logistic regression** — the weights $w_i$ *are* the MaxEnt Lagrange multipliers, fixed uniquely by requiring the model's predicted feature-averages to reproduce the three observed constraints:
$$\sum_{\mathbf x} P(\mathbf x)\, x_i\, \operatorname{sigmoid}(\mathbf w\cdot\mathbf x) = \langle x_i\,\mathbb{I}(Y=1)\rangle, \qquad i=1,2,3.$$

Writing these out scenario-by-scenario (only A, E carry $x_1$; B, D, E carry $x_2$; C, D carry $x_3$):
$$
\begin{aligned}
0.40\,\sigma(w_1) + 0.05\,\sigma(w_1+w_2) &= 0.05 \\
0.25\,\sigma(w_2) + 0.15\,\sigma(w_2+w_3) + 0.05\,\sigma(w_1+w_2) &= 0.15\\
0.05\,\sigma(w_3) + 0.15\,\sigma(w_2+w_3) &= 0.16
\end{aligned}
$$
a nonlinear system in $(w_1,w_2,w_3)$ that we solve numerically below.


```python
import numpy as np
from scipy.optimize import fsolve

def sigmoid(z):
    return 1.0/(1.0+np.exp(-z))

P = {'A':0.40, 'B':0.25, 'C':0.05, 'D':0.15, 'E':0.05}
x = {'A':(1,0,0), 'B':(0,1,0), 'C':(0,0,1), 'D':(0,1,1), 'E':(1,1,0)}
target = {'x1':0.05, 'x2':0.15, 'x3':0.16}

def equations(w):
    w1, w2, w3 = w
    p1 = {k: sigmoid(w1*x[k][0] + w2*x[k][1] + w3*x[k][2]) for k in x}
    c1 = P['A']*x['A'][0]*p1['A'] + P['E']*x['E'][0]*p1['E'] - target['x1']
    c2 = P['B']*x['B'][1]*p1['B'] + P['D']*x['D'][1]*p1['D'] + P['E']*x['E'][1]*p1['E'] - target['x2']
    c3 = P['C']*x['C'][2]*p1['C'] + P['D']*x['D'][2]*p1['D'] - target['x3']
    return [c1, c2, c3]

w_sol, info, ier, msg = fsolve(equations, x0=[0.0, 0.0, 2.0], full_output=True)
print("Converged:", ier==1, "-", msg)
print(f"w1 = {w_sol[0]:.4f}   w2 = {w_sol[1]:.4f}   w3 = {w_sol[2]:.4f}")
print("residuals:", np.round(equations(w_sol), 8))

w1, w2, w3 = w_sol
p1 = {k: sigmoid(w1*x[k][0] + w2*x[k][1] + w3*x[k][2]) for k in x}

print("\nScenario | x            | P(x) | P(Y=1|x)")
for k in x:
    print(f"   {k}     | {x[k]}     | {P[k]:.2f} | {p1[k]:.4f}")
```

## 3.3 Interpretation


The fitted weights come out (see code output) with $w_1 \ll 0$, $w_2<0$ and $w_3\gg0$: financial context strongly *discourages* routing to Technical, account-access context mildly discourages it, and bug/malfunction context strongly *encourages* it — exactly matching the qualitative descriptions given for constraints 1–3. At the pure-technical scenario C, $P(Y=1\mid\mathbf x_C)$ comes out close to but below 1, consistent with the "vast majority but not all" (80% of the $0.20$ ceiling) framing of constraint 3.


```python
import plotly.graph_objects as go

scenario_names = list(x.keys())
probs = [p1[k] for k in scenario_names]
weights = [P[k] for k in scenario_names]

fig = go.Figure()
fig.add_trace(go.Bar(x=scenario_names, y=probs, marker_color='#1f77b4',
                      text=[f"{v:.1%}" for v in probs], textposition='outside',
                      name='P(Y=1 | scenario)'))
fig.update_layout(title='<b>MaxEnt / Logistic Regression: P(Technical | scenario)</b>',
                   xaxis_title='Scenario', yaxis_title='P(Y = 1 | x)',
                   yaxis=dict(range=[0,1.05], tickformat='.0%'),
                   template='plotly_white', width=780, height=460)
fig.show()
```

# Q4. Asset Pricing and Portfolio Optimization

## 4.1 Setup


We want the risk-neutral density $p(S_T)$ for $S_T>0$ of Asset X one month out, given three market-implied moments:
$$\mathbb{E}[S_T] = 102, \qquad \mathbb{E}[S_T^2] = 10{,}600, \qquad \mathbb{E}[\ln S_T] = 4.60.$$

We maximize the continuous entropy $H_c[p] = -\int_0^\infty p(S)\ln p(S)\,dS$ (flat reference measure $m(S)=1$ on $S>0$) subject to normalization and the three constraints above.


## 4.2 MaxEnt Form


With Lagrange multipliers $\lambda_1,\lambda_2,\lambda_3$ for the mean, second-moment and log-mean constraints respectively, the same functional-derivative argument as in Q2 gives
$$p^*(S) = \frac1Z\exp\left(-\lambda_1 S - \lambda_2 S^2 - \lambda_3\ln S\right)
= \frac{1}{Z}\, S^{-\lambda_3}\, \exp\left(-\lambda_1 S - \lambda_2 S^2\right), \qquad S>0.$$

This is a **generalized Gaussian-power-law hybrid**: the $S^2$ term in the exponent gives Gaussian-like decay for large $S$ (controlling the variance), while the extra $S^{-\lambda_3}$ power-law factor reshapes the density near the origin, producing the negative skew ("fat downside tail") that the $\mathbb{E}[\ln S_T]$ constraint is designed to encode — exactly the effect described in the problem statement (put-driven downside sentiment). There is no closed form for $Z$ or for $(\lambda_1,\lambda_2,\lambda_3)$ in terms of $(102, 10600, 4.60)$, so we solve for the multipliers numerically by matching moments.


```python
import numpy as np
from scipy.optimize import least_squares
from scipy.integrate import quad
import warnings
warnings.filterwarnings("ignore")

target_mean, target_sq, target_log = 102.0, 10600.0, 4.60

def peak_and_width(l1, l2, l3):
    l2 = max(l2, 1e-8)
    a, b, c = 2*l2, l1, -l3          # root of d(log g)/dS = 0
    disc = b*b - 4*a*c
    if disc < 0:
        Speak = 100.0
    else:
        roots = [(-b + s*np.sqrt(disc))/(2*a) for s in (1, -1)]
        cands = [r for r in roots if r > 0]
        Speak = max(cands) if cands else 100.0
    curv = abs(l3/(Speak**2) + 2*l2)
    width = 1/np.sqrt(curv) if curv > 1e-10 else 50.0
    return Speak, width

def moments(lam):
    l1, l2, l3 = lam
    l2 = max(l2, 1e-8)
    Speak, width = peak_and_width(l1, l2, l3)
    lo, hi = max(1e-4, Speak - 12*width), Speak + 12*width
    def logg(S):
        return -l3*np.log(S) - l1*S - l2*S**2
    lgmax = logg(Speak)
    g = lambda S: np.exp(logg(S) - lgmax)
    Z,_    = quad(g, lo, hi, limit=400)
    m1,_   = quad(lambda S: S*g(S), lo, hi, limit=400)
    m2,_   = quad(lambda S: S**2*g(S), lo, hi, limit=400)
    mlog,_ = quad(lambda S: np.log(S)*g(S), lo, hi, limit=400)
    return Z, m1/Z, m2/Z, mlog/Z

def residuals(lam):
    Z, E1, E2, Elog = moments(lam)
    return [E1-target_mean, (E2-target_sq)/100.0, (Elog-target_log)*10]

best = None
for l2_0 in [0.0005, 0.001, 0.00255, 0.005, 0.01]:
    for l3_0 in [-5, -1, 0, 1, 5, 20, 50]:
        l1_0 = -2*l2_0*target_mean - l3_0/target_mean
        try:
            sol = least_squares(residuals, [l1_0, l2_0, l3_0],
                                 bounds=([-10, 1e-6, -500], [10, 1, 500]))
            cost = sum(v*v for v in residuals(sol.x))
            if best is None or cost < best[0]:
                best = (cost, sol.x)
        except Exception:
            pass

lam1, lam2, lam3 = best[1]
Z, E1, E2, Elog = moments(best[1])
print(f"lambda_1 = {lam1:.6f}")
print(f"lambda_2 = {lam2:.6f}")
print(f"lambda_3 = {lam3:.6f}")
print(f"Z        = {Z:.6f}")
print()
print(f"Check  E[S]    = {E1:.4f}  (target 102.00)")
print(f"Check  E[S^2]  = {E2:.4f}  (target 10600.00)")
print(f"Check  E[ln S] = {Elog:.4f}  (target 4.60)")
print(f"Implied Var[S] = {E2-E1**2:.4f}   Std[S] = {np.sqrt(E2-E1**2):.4f}")
```

## 4.3 Result


The numerical solve (see output above) converges to
$$\lambda_1 \approx -0.9547, \qquad \lambda_2 \approx 0.00408, \qquad \lambda_3 \approx 11.94, \qquad Z\approx 319.29,$$
so the least-biased risk-neutral density consistent with the market's forward price, quadratic-variation and log-price constraints is
$$\boxed{p^*(S) = \frac{1}{319.29}\, S^{-11.94}\, \exp\left(0.9547\,S - 0.00408\,S^2\right), \qquad S>0.}$$

The resulting distribution has mean \$102, standard deviation $\approx\$14$, and — because of the $S^{-\lambda_3}$ term — a mode below the mean and a heavier left tail than a Normal$(102,14^2)$ would have, reflecting the market's priced-in downside risk without having to assume any particular parametric model (e.g. lognormal / geometric Brownian motion) up front.


```python
import plotly.graph_objects as go
import numpy as np

S = np.linspace(50, 160, 600)
logg = -lam3*np.log(S) - lam1*S - lam2*S**2
p = np.exp(logg - logg.max())
p = p / np.trapezoid(p, S)     # normalize on the plotting grid for display

# comparison Normal(102, 14^2) with matching mean/variance only
from scipy.stats import norm
p_gauss = norm.pdf(S, loc=E1, scale=np.sqrt(E2-E1**2))

fig = go.Figure()
fig.add_trace(go.Scatter(x=S, y=p, mode='lines', name='MaxEnt risk-neutral density  p*(S)',
                          line=dict(color='#1f77b4', width=3)))
fig.add_trace(go.Scatter(x=S, y=p_gauss, mode='lines', name='Normal(102, 14²) (mean/var only)',
                          line=dict(color='#d62728', width=2, dash='dash')))
fig.add_vline(x=102, line=dict(color='gray', dash='dot'), annotation_text='E[S]=102')

fig.update_layout(title='<b>MaxEnt Risk-Neutral Density for Asset X, S_T</b>',
                   xaxis_title='S_T ($)', yaxis_title='Probability density',
                   template='plotly_white', width=850, height=500)
fig.show()
```

## 4.4 Why This Matters


Notice the MaxEnt density is visibly **left-skewed relative to the variance-matched Normal** (dashed red curve): it shifts probability mass from the right tail into a slightly heavier left tail while keeping the same mean and (nearly) the same variance, purely because the market's $\mathbb{E}[\ln S_T]=4.60$ constraint is *below* what a symmetric Normal with the same mean/variance would imply. This is precisely the "fat tails / skew" behavior described in the problem: MaxEnt lets the *option-implied* moment (constraint 3) reshape the distribution's skewness without ever assuming a parametric form like geometric Brownian motion — it is the least-biased density consistent with exactly the market data observed, no more and no less.
