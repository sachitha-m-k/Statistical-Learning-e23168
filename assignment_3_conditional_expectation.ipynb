## Q1.3: Why Must $E[X|Y]$ be $\sigma(Y)$-Measurable?

**Answer:**

**Theoretical Justification:**

By definition, the conditional expectation $E[X|Y]$ is the unique random variable that satisfies:

1. **Measurability:** $E[X|Y]$ is $\sigma(Y)$-measurable.
2. **Integral Property:** For all $A \in \sigma(Y)$: $\int_A E[X|Y] \, dP = \int_A X \, dP$

By the **Doob-Dynkin Lemma**, any $\sigma(Y)$-measurable random variable must be a function of $Y$. Therefore:
$$E[X|Y] = g(Y)$$

for some measurable function $g$.

**Implication for $\sigma(Y)$-Constant Sets:**

Since $E[X|Y]$ is $\sigma(Y)$-measurable, it must be constant on each element of $\sigma(Y)$.

- On $\{\omega_1, \omega_2\}$: $E[X|Y](\omega_1) = E[X|Y](\omega_2) = c_0$ (constant)
- On $\{\omega_3, \omega_4\}$: $E[X|Y](\omega_3) = E[X|Y](\omega_4) = c_1$ (constant)

**Physical Interpretation:**

The bank can only observe whether a customer is low-risk or high-risk. Since $Y$ is constant on $\{\omega_1, \omega_2\}$, the bank **cannot distinguish** between legitimate transactions from low-risk customers and fraudulent transactions from low-risk customers based on $Y$ alone. Thus, the best estimate of fraud probability must be the same for both outcomes in that set.

## Q1.4: Compute $E[X|Y]$ Explicitly

For $A = \{\omega_1, \omega_2\}$ (the set where $Y=0$):

$$\int_A E[X|Y] \, dP = \int_A X \, dP$$

$$c_0 \cdot P(A) = X(\omega_1) P(\{\omega_1\}) + X(\omega_2) P(\{\omega_2\})$$

$$c_0 \cdot (0.50 + 0.10) = 0 \cdot 0.50 + 1 \cdot 0.10$$

$$c_0 \cdot 0.60 = 0.10$$

$$c_0 = \frac{0.10}{0.60} = \frac{1}{6}$$

$$c_1 \cdot 0.40 = 0.20$$

**Result:**

## Q1.6: Find the Conditional PMF $p_{X|Y}(x|y)$

**Answer:**

The conditional probability mass function is defined as:
$$p_{X|Y}(x|y) = P(X=x|Y=y) = \frac{P(X=x, Y=y)}{P(Y=y)}$$

**Step 1: Compute marginal probabilities of $Y$**

$$P(Y=0) = P(\{\omega_1, \omega_2\}) = 0.50 + 0.10 = 0.60$$
$$P(Y=1) = P(\{\omega_3, \omega_4\}) = 0.20 + 0.20 = 0.40$$

**Step 2: Compute joint probabilities**

The joint distribution is:

| $(X, Y)$ | Probability |
|:---:|:---:|
| $(0, 0)$ | $P(X=0, Y=0) = P(\{\omega_1, \omega_2\}) = 0.50 + 0.10 = 0.60$ |
| $(1, 0)$ | $P(X=1, Y=0) = P(\{\omega_2\}) = 0.10$ |
| $(0, 1)$ | $P(X=0, Y=1) = P(\{\omega_3\}) = 0.20$ |
| $(1, 1)$ | $P(X=1, Y=1) = P(\{\omega_4\}) = 0.20$ |

**Step 3: Apply the conditional probability formula**

**For $Y=0$:**
$$p_{X|Y}(0|0) = \frac{P(X=0, Y=0)}{P(Y=0)} = \frac{0.50}{0.60} = \frac{5}{6}$$
$$p_{X|Y}(1|0) = \frac{P(X=1, Y=0)}{P(Y=0)} = \frac{0.10}{0.60} = \frac{1}{6}$$

**For $Y=1$:**
$$p_{X|Y}(0|1) = \frac{P(X=0, Y=1)}{P(Y=1)} = \frac{0.20}{0.40} = \frac{1}{2}$$
$$p_{X|Y}(1|1) = \frac{P(X=1, Y=1)}{P(Y=1)} = \frac{0.20}{0.40} = \frac{1}{2}$$

**Result:**

$$p_{X|Y}(x|y) = \begin{cases}
\frac{5}{6} & \text{if } x=0, y=0 \\
\frac{1}{6} & \text{if } x=1, y=0 \\
\frac{1}{2} & \text{if } x=0, y=1 \\
\frac{1}{2} & \text{if } x=1, y=1

**Table Representation:**

| $p_{X\|Y}(x\|y)$ | $y=0$ | $y=1$ |
|:---:|:---:|:---:|

## Q1.10: Information Theory - Entropy Analysis

**Answer:**

### Q1.10(a): Compute the Entropy $H(X)$

**Definition:** Information entropy (also called Shannon entropy) for a discrete random variable:
$$H(X) = -\sum_x P(X=x) \log_2 P(X=x)$$

**Marginal distribution of $X$:**
- $P(X=0) = P(\{\omega_1, \omega_3\}) = 0.50 + 0.20 = 0.70$
- $P(X=1) = P(\{\omega_2, \omega_4\}) = 0.10 + 0.20 = 0.30$

**Calculate:**
$$H(X) = -[0.70 \log_2(0.70) + 0.30 \log_2(0.30)]$$
$$= -[0.70 \cdot (-0.5145) + 0.30 \cdot (-1.7370)]$$
$$= -[-0.3602 - 0.5211]$$
$$= -[-0.8813] = 0.8813 \text{ bits}$$

**Result:**
$$\boxed{H(X) \approx 0.881 \text{ bits}}$$

**Interpretation:** The bank initially has about 0.881 bits of uncertainty about whether a transaction is fraudulent.

---

### Q1.10(b): Compute the Conditional Entropy $H(X|Y)$

**Definition:**
$$H(X|Y) = E[H(X|Y=y)] = \sum_y P(Y=y) H(X|Y=y)$$

where
$$H(X|Y=y) = -\sum_x p_{X|Y}(x|y) \log_2 p_{X|Y}(x|y)$$

**For $y=0$ (low-risk customers):**
$$H(X|Y=0) = -[\frac{5}{6} \log_2(\frac{5}{6}) + \frac{1}{6} \log_2(\frac{1}{6})]$$
$$= -[\frac{5}{6} \cdot (-0.2630) + \frac{1}{6} \cdot (-2.5850)]$$
$$= -[-0.2192 - 0.4308] = 0.6500 \text{ bits}$$

**For $y=1$ (high-risk customers):**
$$H(X|Y=1) = -[\frac{1}{2} \log_2(\frac{1}{2}) + \frac{1}{2} \log_2(\frac{1}{2})]$$
$$= -[\frac{1}{2} \cdot (-1) + \frac{1}{2} \cdot (-1)]$$
$$= -[-1] = 1.0 \text{ bit}$$

**Weighted average:**
$$H(X|Y) = 0.60 \cdot 0.6500 + 0.40 \cdot 1.0 = 0.39 + 0.40 = 0.79 \text{ bits}$$

**Result:**
$$\boxed{H(X|Y) \approx 0.790 \text{ bits}}$$

**Interpretation:** After knowing whether a customer is low-risk or high-risk, the remaining uncertainty about fraud is about 0.790 bits.

---

### Q1.10(c): Calculate the Mutual Information $I(X; Y)$

**Definition:**
$$I(X; Y) = H(X) - H(X|Y)$$

**Calculate:**
$$I(X; Y) = 0.881 - 0.790 = 0.091 \text{ bits}$$

**Alternative calculation using symmetry:**

**Marginal distribution of $Y$:**
- $P(Y=0) = 0.60$
- $P(Y=1) = 0.40$

$$= -[0.60 \cdot (-0.7370) + 0.40 \cdot (-1.3219)]$$
$$= 0.9710 \text{ bits}$$

**Result:**
$$\boxed{I(X; Y) \approx 0.091 \text{ bits}}$$

---

### Q1.10(d): Percentage of Uncertainty Removed

**Calculation:**

$$= \frac{0.091}{0.881} \times 100\% = 10.3\%$$

**Result:**

# Q2: Conditional Density of the Multivariate Gaussian

## Q2: Deriving the Conditional Density for Multivariate Gaussians

**Answer:**

### Proof Strategy

We start with the joint density of $(\mathbf{X}, \mathbf{Y})$:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right)$$

where $\mathbf{z} = \begin{bmatrix} \mathbf{x} \\ \mathbf{y} \end{bmatrix}$ and $\boldsymbol{\mu} = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix}$

### Step 1: Partition the Covariance Matrix Inverse

Using block matrix inversion, we can write:

$$\boldsymbol{\Sigma}^{-1} = \begin{bmatrix} \mathbf{V}_{XX} & \mathbf{V}_{XY} \\ \mathbf{V}_{YX} & \mathbf{V}_{YY} \end{bmatrix}$$

where

$$\mathbf{V}_{XX} = \left(\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}\right)^{-1} = \boldsymbol{\Sigma}_{X|Y}^{-1}$$
$$\mathbf{V}_{YY} = \left(\boldsymbol{\Sigma}_{YY} - \boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{XX}^{-1}\boldsymbol{\Sigma}_{XY}\right)^{-1}$$

### Step 2: Expand the Quadratic Form

The key insight is to decompose the quadratic form in the exponent:

$$\begin{align}
(\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) &= (\mathbf{x} - \boldsymbol{\mu}_X - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y))^T \times \\
&\quad \boldsymbol{\Sigma}_{X|Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}_X - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)) \\
&\quad + (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y)
\end{align}$$

### Step 3: Separate the Joint Density

The joint density can be written as:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y}) = f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x}|\mathbf{y}) \cdot f_\mathbf{Y}(\mathbf{y})$$

### Step 4: Extract the Conditional Density

The conditional density is:

$$\boxed{f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{X|Y}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X|Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right)}$$

where

- **Conditional Covariance:**
$$\boldsymbol{\Sigma}_{X|Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$$

- **Conditional Mean:**
$$\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$$

### Conclusion

**Key Result:**
$$\mathbf{X} | \mathbf{Y}=\mathbf{y} \sim \mathscr{N}(\boldsymbol{\mu}^*, \boldsymbol{\Sigma}_{X|Y})$$

The conditional distribution of $\mathbf{X}$ given $\mathbf{Y}=\mathbf{y}$ is **also Gaussian** with:
- Mean: $\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$
- Covariance: $\boldsymbol{\Sigma}_{X|Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$

## Q3.1: Joint Distribution

**Answer:**

### Part A: Show that $Y \sim \mathscr{N}(0, 5)$

**Given:**
- $X \sim \mathscr{N}(0, 1)$
- $Q \sim \mathscr{N}(0, 4)$
- $X$ and $Q$ are independent
- $Y = X + Q$

**Solution:**

Since $Y$ is a linear combination of independent normal random variables, $Y$ is also normal.

**Mean:**
$$E[Y] = E[X + Q] = E[X] + E[Q] = 0 + 0 = 0$$

**Variance:**
Since $X$ and $Q$ are independent:
$$\mathrm{Var}(Y) = \mathrm{Var}(X + Q) = \mathrm{Var}(X) + \mathrm{Var}(Q) = 1 + 4 = 5$$

**Result:**
$$\boxed{Y \sim \mathscr{N}(0, 5)}$$

---

### Part B: Find the Covariance $\text{Cov}(X, Y)$

**Solution:**

$$\text{Cov}(X, Y) = \text{Cov}(X, X + Q)$$

$$= \text{Cov}(X, X) + \text{Cov}(X, Q)$$

Since $X$ and $Q$ are independent: $\text{Cov}(X, Q) = 0$

$$= \mathrm{Var}(X) + 0 = 1$$

**Result:**
$$\boxed{\text{Cov}(X, Y) = 1}$$

---

### Summary: Joint Distribution Parameters

$$\begin{bmatrix} X \\ Y \end{bmatrix} \sim \mathscr{N} \left( \begin{bmatrix} 0 \\ 0 \end{bmatrix}, \begin{bmatrix} 1 & 1 \\ 1 & 5 \end{bmatrix} \right)$$

## Q3.2: Information Geometry - Why $X$ is Not $\sigma(Y)$-Measurable

**Answer:**

### Theoretical Explanation

**Definition:** A random variable $X$ is $\sigma(Y)$-measurable if $X$ is a deterministic function of $Y$. That is, $X = g(Y)$ for some measurable function $g$.

**Why $X$ is NOT $\sigma(Y)$-measurable:**

Consider the measure-theoretic definition: For $X$ to be $\sigma(Y)$-measurable, for every $t \in \mathbb{R}$, the set $\{\omega: X(\omega) \leq t\}$ must be expressible in terms of $Y$. In other words, if $Y(\omega_1) = Y(\omega_2)$, then we must have $X(\omega_1) = X(\omega_2)$.

In our case, $Y = X + Q$, so:
- We can have $Y(\omega_1) = Y(\omega_2)$ for outcomes with different $X$ values
- This happens when $X_1 + Q_1 = X_2 + Q_2$ but $X_1 \neq X_2$
- Therefore, $X$ cannot be expressed as a function of $Y$ alone

### Practical Interpretation

**The Engineer's Dilemma:**

Suppose the power meter shows $Y = 5$. The engineer observes:

$$Y = X + Q = 5$$

**Two scenarios could produce this reading:**

1. **Motor is overheating** ($X$ is high):
   - $X = 4, Q = 1$ → $Y = 5$ ✓
   - High temperature + low mechanical load

2. **Material is very hard** ($Q$ is high, $X$ is low):
   - $X = 1, Q = 4$ → $Y = 5$ ✓
   - Low temperature + high mechanical load

**The Fundamental Problem:**

Just by looking at the power reading ($Y=5$), the engineer **cannot determine with certainty** whether:
- The motor is overheating (a serious problem requiring immediate action)
- The cutting material is simply very hard (a normal operational condition)

This is precisely what "$X$ is not $\sigma(Y)$-measurable" means in practical terms.

**Mathematical Formalization:**

The set of possible values of $X$ given $Y=5$:
$$X | Y=5 \in (-\infty, 5)$$

Any value from negative infinity up to 5 is possible, depending on the unobserved mechanical load $Q = 5 - X$.

## Q3.3: Conditional Expectation (The Best Estimate)

**Answer:**

### Computing $E[X|Y]$ Using Gaussian Properties

For a bivariate normal distribution, the conditional expectation is:
$$E[X|Y=y] = \mu_X + \frac{\text{Cov}(X,Y)}{\mathrm{Var}(Y)}(y - \mu_Y)$$

**Substituting our parameters:**
- $\mu_X = 0$
- $\mu_Y = 0$
- $\text{Cov}(X,Y) = 1$
- $\mathrm{Var}(Y) = 5$

$$E[X|Y=y] = 0 + \frac{1}{5}(y - 0) = \frac{1}{5}y$$

**As a Random Variable:**
$$\boxed{E[X|Y] = \frac{1}{5}Y}$$

---

### Interpretation as Signal-to-Total-Variance Ratio

We can rewrite the conditional expectation in a more intuitive form:

$$E[X|Y] = \frac{\text{Cov}(X,Y)}{\mathrm{Var}(Y)} Y = \frac{\mathrm{Var}(X)}{\mathrm{Var}(X) + \mathrm{Var}(Q)} Y$$

$$E[X|Y] = \frac{1}{1+4} Y = \frac{1}{5}Y$$

**Physical Interpretation:**

The fraction $\frac{1}{5}$ represents how much of the observed power $Y$ should be attributed to the actual temperature $X$:

- **Numerator ($\mathrm{Var}(X) = 1$):** The variance of the true signal (temperature)
- **Denominator ($\mathrm{Var}(X) + \mathrm{Var}(Q) = 5$):** The total variance (signal + noise)

The ratio $\frac{1}{5}$ is the **relative signal strength**. Since the mechanical noise has much larger variance (4) than the temperature signal (1), the measured power is dominated by noise. Therefore, when observing $Y=y$, we only attribute $\frac{y}{5}$ to the true temperature.

---

### Practical Example

If the power meter reads $Y = 5$ units:
$$E[X|Y=5] = \frac{1}{5} \cdot 5 = 1 \text{ unit}$$

The engineer estimates the temperature at 1 unit, **not** 5 units. This is the "shrinkage" toward zero caused by the noise.

## Q3.4: Verification of the Residual (Orthogonality Principle)

**Answer:**

**Definition:** Let $\widehat{X} = E[X|Y]$ be our estimate, and define the residual:
$$Z = X - \widehat{X} = X - E[X|Y]$$

---

### Part A: Prove that $E[Z] = 0$

**Proof:**

Using the law of iterated expectations:
$$E[Z] = E[X - E[X|Y]] = E[X] - E[E[X|Y]]$$

By the law of total expectation:
$$E[E[X|Y]] = E[X]$$

Therefore:
$$E[Z] = E[X] - E[X] = 0 \quad \checkmark$$

---

### Part B: Verify Orthogonality - $\text{Cov}(Z, Y) = 0$

**The Orthogonality Principle states:** The estimation error $Z$ is uncorrelated with the observation $Y$.

**Proof:**

$$\text{Cov}(Z, Y) = \text{Cov}(X - E[X|Y], Y)$$

$$= \text{Cov}(X, Y) - \text{Cov}(E[X|Y], Y)$$

**Key fact:** $E[X|Y]$ is a function of $Y$, say $E[X|Y] = g(Y)$. Therefore:

$$\text{Cov}(E[X|Y], Y) = \text{Cov}(g(Y), Y)$$

For our linear case:
$$\text{Cov}(E[X|Y], Y) = \text{Cov}(\frac{1}{5}Y, Y) = \frac{1}{5}\mathrm{Var}(Y) = \frac{1}{5} \cdot 5 = 1$$

**Therefore:**
$$\text{Cov}(Z, Y) = \text{Cov}(X, Y) - \text{Cov}(E[X|Y], Y) = 1 - 1 = 0 \quad \checkmark$$

---

### Physical Interpretation

**The orthogonality principle reveals:**

The error $Z = X - E[X|Y]$ contains **no information about the observation** $Y$. In other words:

- The useful information in $Y$ has been fully extracted by $E[X|Y]$
- Any remaining error is purely noise/randomness unrelated to the measurement
- This confirms that $E[X|Y]$ is the optimal estimate: it squeezes all possible information from $Y$

## Q3.5: Information Theory - Differential Entropy

**Answer:**

### Part A: Differential Entropy $h(X)$

**Definition:** For a continuous random variable with density $f(x)$:
$$h(X) = -\int_{-\infty}^{\infty} f(x) \log_2 f(x) \, dx$$

**For a Gaussian distribution $X \sim \mathscr{N}(\mu, \sigma^2)$:**
$$h(X) = \frac{1}{2}\log_2(2\pi e \sigma^2) \text{ (in bits)}$$

**For our case:** $X \sim \mathscr{N}(0, 1)$
$$h(X) = \frac{1}{2}\log_2(2\pi e \cdot 1)$$
$$= \frac{1}{2}\log_2(2\pi e)$$
$$= \frac{1}{2} \cdot 4.8534 = 2.426 \text{ bits}$$

**Result:**
$$\boxed{h(X) \approx 2.426 \text{ bits}}$$

---

### Part B: Conditional Differential Entropy $h(X|Y)$

**For Gaussian distributions:**
$$h(X|Y) = \frac{1}{2}\log_2(2\pi e \sigma_{X|Y}^2)$$

**Step 1: Compute the conditional variance**

From the theory of bivariate normals:
$$\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\mathrm{Var}(Y)}$$

$$= 1 - \frac{1^2}{5} = 1 - 0.2 = 0.8$$

**Step 2: Calculate the conditional entropy**
$$h(X|Y) = \frac{1}{2}\log_2(2\pi e \cdot 0.8)$$
$$= \frac{1}{2}\log_2(5.0689)$$
$$= \frac{1}{2} \cdot 2.3425 = 1.171 \text{ bits}$$

**Result:**
$$\boxed{h(X|Y) \approx 1.171 \text{ bits}}$$

---

### Physical Interpretation

- **Before measurement:** The engineer has $h(X) \approx 2.426$ bits of uncertainty
- **After measurement:** The engineer still has $h(X|Y) \approx 1.171$ bits of uncertainty
- **Information gained:** About $2.426 - 1.171 = 1.255$ bits

## Q3.6: Mutual Information

**Answer:**

### Part A: Calculate $I(X; Y)$

**Definition:**
$$I(X; Y) = h(X) - h(X|Y)$$

**Calculation:**
$$I(X; Y) = 2.426 - 1.171 = 1.255 \text{ bits}$$

**Result:**
$$\boxed{I(X; Y) \approx 1.255 \text{ bits}}$$

---

### Part B: Percentage of Uncertainty Removed

$$\text{Percentage Removed} = \frac{I(X; Y)}{h(X)} \times 100\%$$
$$= \frac{1.255}{2.426} \times 100\% \approx 51.8\%$$

**Result:**
$$\boxed{\text{About 51.8\% of the uncertainty is removed}}$$

**Interpretation:** Observing the power meter removes about **half** of the engineer's initial uncertainty about the true motor temperature. This is a substantial information gain.

---

### Part C: Impact of Increased Mechanical Noise

**Scenario:** If $Q \sim \mathscr{N}(0, 100)$ instead of $\mathscr{N}(0, 4)$

**New parameters:**
- $\mathrm{Var}(Y) = \mathrm{Var}(X) + \mathrm{Var}(Q) = 1 + 100 = 101$
- $\text{Cov}(X, Y) = 1$ (unchanged)
- $\sigma_{X|Y}^2 = 1 - \frac{1}{101} \approx 0.9901$

**New conditional entropy:**
$$h(X|Y) = \frac{1}{2}\log_2(2\pi e \cdot 0.9901) \approx 2.419 \text{ bits}$$

**New mutual information:**
$$I(X; Y) = 2.426 - 2.419 = 0.007 \text{ bits}$$

**Percentage removed:**
$$\frac{0.007}{2.426} \times 100\% \approx 0.3\%$$

**Dramatic Change:**
- **With low noise ($\mathrm{Var}(Q)=4$):** Power meter removes **51.8%** of uncertainty
- **With high noise ($\mathrm{Var}(Q)=100$):** Power meter removes only **0.3%** of uncertainty

**Conclusion:** When mechanical noise dominates (100x larger than temperature variance), the power meter becomes nearly useless for inferring temperature. The engineer would need other sensors (e.g., direct temperature probes) to monitor the motor's health.

## Q3.7: Diagnostic Interpretation - Why $Y$ is Biased and Shrinkage to the Mean

**Answer:**

### The Manager's Claim

The factory manager suggests: *"Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself."*

This reasoning seems intuitive but is fundamentally flawed from a statistical perspective.

---

### Part A: Why $Y$ is a BIASED Estimator of $X$

**Definition:** An estimator $\widehat{X}$ is unbiased if $E[\widehat{X}] = E[X]$.

**Check if $\widehat{X} = Y$ is unbiased:**
$$E[Y] = E[X + Q] = E[X] + E[Q] = 0 + 0 = 0 = E[X]$$

Actually, $Y$ **is unbiased** in expectation. However, this is not the right criterion.

**The Real Problem: Mean Square Error (MSE)**

The optimal estimator minimizes MSE, not just bias.

$$MSE(Y) = E[(X - Y)^2] = E[(X - X - Q)^2] = E[Q^2] = \mathrm{Var}(Q) = 4$$

Compare with our estimate from Q3.3:
$$MSE(E[X|Y]) = E[(X - E[X|Y])^2] = \sigma_{X|Y}^2 = 0.8$$

**Result:**
$$MSE(E[X|Y]) = 0.8 < 4 = MSE(Y)$$

**Conclusion:** Although $Y$ is unbiased, it's a much worse estimator than $E[X|Y]$ because it has **higher MSE**. The unbiasedness is offset by huge variance (noise).

---

### Part B: Shrinkage Toward the Mean

**Why does $E[X|Y]$ "shrink" toward zero?**

Recall:
$$E[X|Y=y] = \frac{1}{5}y$$

For example:
- If $Y = 5$, we estimate $E[X|Y=5] = 1$ (shrunk from 5 to 1)
- If $Y = -10$, we estimate $E[X|Y=-10] = -2$ (shrunk from -10 to -2)

**Why does this shrinkage happen?**

The measurement $Y = X + Q$ contains:
1. **Signal** (the true temperature $X$): variance = 1
2. **Noise** (mechanical load $Q$): variance = 4

The ratio of signal to total is:
$$\frac{\text{Signal variance}}{\text{Total variance}} = \frac{1}{1+4} = \frac{1}{5}$$

**Statistical Principle:** In the presence of noise, we should "believe" the measurement only to the extent that the true signal dominates. Here, the true signal is only $\frac{1}{5}$ of the total variance, so we trust the measurement by only $\frac{1}{5}$.

---

### Part C: Formal Proof of Shrinkage

The conditional expectation can be rewritten as:
$$E[X|Y] = E[X] + \frac{\text{Cov}(X,Y)}{\mathrm{Var}(Y)}(Y - E[Y])$$
$$= 0 + \frac{1}{5}Y = \frac{1}{5}Y$$

This is a **weighted average** of:
- **Prior expectation:** $E[X] = 0$ (with weight $1 - \frac{1}{5} = \frac{4}{5}$)
- **Measurement:** $Y$ (with weight $\frac{1}{5}$)

$$E[X|Y] = \frac{4}{5} \cdot 0 + \frac{1}{5} \cdot Y = \frac{1}{5}Y$$

The large weight on the prior ($\frac{4}{5}$) reflects our strong belief that the measurement is mostly noise.

---

### Part D: Practical Engineering Insight

**Why the manager's intuition fails:**

The manager sees the simple equation $Y = X + Q$ and thinks $X = Y$ is the natural answer. But this ignores the **noise magnitude**.

**The Kalman Filter perspective:**

$E[X|Y]$ is precisely what a **Kalman Filter** would compute:

1. **Prior:** Assume $E[X] = 0$ (from historical data)
2. **Measurement:** Observe $Y$
3. **Update:** $E[X|Y] = E[X] + \text{Gain} \times (Y - E[Y])$
   - Gain = $\frac{\text{Signal Variance}}{\text{Signal + Noise Variance}}$
   - Gain = $\frac{1}{5}$
4. **Result:** $E[X|Y] = \frac{1}{5}Y$

This is the **optimal** way to blend prior knowledge with noisy measurements.
